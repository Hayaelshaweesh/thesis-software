```python
def impute_mode(df_missing):
    """
    Optimized mode imputation using scikit-learn
    - Faster for large DataFrames
    - No truth value ambiguity
    """
    try:
        # Create separate imputers for numeric and categorical data
        numeric_cols = df_missing.select_dtypes(include=np.number).columns
        categ_cols = df_missing.select_dtypes(exclude=np.number).columns

        df_imputed = df_missing.copy()

        if len(numeric_cols) > 0:
            num_imputer = SimpleImputer(strategy='most_frequent')
            df_imputed[numeric_cols] = num_imputer.fit_transform(df_imputed[numeric_cols])

        if len(categ_cols) > 0:
            cat_imputer = SimpleImputer(strategy='most_frequent')
            df_imputed[categ_cols] = cat_imputer.fit_transform(df_imputed[categ_cols])

        return df_imputed

    except Exception as e:
        print(f"Error in optimized impute_mode: {str(e)}")
        return df_missing.copy()

def impute_knn(df_missing, n_neighbors=3):
    """
    KNN imputation with ordinal encoding
    Uses the values from the k nearest neighbors (based on other variables) to impute the missing value.
"""

    try:
        # Input validation
        if not isinstance(df_missing, pd.DataFrame):
            raise ValueError("Input must be a pandas DataFrame")
        if df_missing.empty:
            raise ValueError("DataFrame cannot be empty")
        if n_neighbors < 1:
            raise ValueError("n_neighbors must be at least 1")
         # Ordinal encoding

        encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

        ##3 encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
        try:
            encoded = encoder.fit_transform(df_missing)
        except Exception as e:
            raise ValueError(f"Ordinal encoding failed: {str(e)}")

        # KNN imputation
        imputer = KNNImputer(n_neighbors=n_neighbors)
        try:
            imputed_encoded = imputer.fit_transform(encoded)
        except Exception as e:
            raise ValueError(f"KNN imputation failed: {str(e)}")

        # Inverse transform
        try:
            result = pd.DataFrame(encoder.inverse_transform(imputed_encoded),
                                columns=df_missing.columns)
        except Exception as e:
            raise ValueError(f"Inverse transform failed: {str(e)}")

        return result

    except Exception as e:
        # Catch any unexpected errors
        raise RuntimeError(f"Imputation failed: {str(e)}") from e



def impute_knn_one_column_at_a_time(df_missing, n_neighbors=3):
    # Process one column at a time if memory constrained
    imputed = df_missing.copy()
    for col in df_missing.columns:
        if df_missing[col].isna().any():
            # Only encode/impute one column at a time
            col_data = df_missing[[col]].copy()

            encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
            encoded = encoder.fit_transform(col_data)
            imputer = KNNImputer(n_neighbors=n_neighbors)
            imputed_col = imputer.fit_transform(encoded)
            imputed[col] = encoder.inverse_transform(imputed_col).ravel()
    return imputed

def impute_iterative(df_missing):
    """Iterative/MICE imputation"""
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)


##9    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
    encoded = encoder.fit_transform(df_missing)
    imputer = IterativeImputer(random_state=42, max_iter=10)
    imputed_encoded = imputer.fit_transform(encoded)
    imputed_encoded = np.where(imputed_encoded > encoder.categories_[0].shape[0] - 1, -1, imputed_encoded)
    imputed_encoded = imputed_encoded.astype(int)
    return pd.DataFrame(encoder.inverse_transform(imputed_encoded), columns=df_missing.columns)
```

```python
def impute_hot_deck(df_missing: pd.DataFrame, donor_pool_size: int = 5, random_state: int = 42) -> pd.DataFrame:
    """
    Hot Deck imputation for categorical data.

    For each record with missing values, finds similar donor records from complete cases
    and randomly selects one to impute missing values.

    """
    if not isinstance(df_missing, pd.DataFrame):
        raise ValueError("Input must be a pandas DataFrame")
    if df_missing.empty:
        return df_missing.copy()

    # Set random seed
    np.random.seed(random_state)

    # Create copy to avoid modifying original
    df_imputed = df_missing.copy()

    # Get complete cases (potential donors)
    complete_mask = ~df_missing.isna().any(axis=1)
    donors = df_missing[complete_mask]

    # If no complete cases, fall back to mode imputation
    if len(donors) == 0:
        print("Warning: No complete cases found. Falling back to mode imputation.")
        return impute_mode(df_missing)

    # For each record with missing values
    for idx, row in df_missing[~complete_mask].iterrows():
        # Find matching donors based on observed values
        observed_mask = ~row.isna()
        observed_cols = row.index[observed_mask]

        if len(observed_cols) > 0:
            # Calculate similarity (number of matching values)
            similarity = (donors[observed_cols] == row[observed_cols]).sum(axis=1)
            potential_donors = donors[similarity == similarity.max()]

            # Randomly select from top similar donors
            if len(potential_donors) > donor_pool_size:
                potential_donors = potential_donors.sample(donor_pool_size, random_state=random_state)

            # Select random donor
            donor = potential_donors.sample(1, random_state=random_state).iloc[0]
        else:
            # If no observed values, select random donor
            donor = donors.sample(1, random_state=random_state).iloc[0]

        # Impute missing values from donor
        missing_cols = row.index[row.isna()]
        df_imputed.loc[idx, missing_cols] = donor[missing_cols]

    return df_imputed
```

```python
def robust_mice_impute(df, max_iter=10, random_state=42):
    """
    MICE imputation that handles single-class categorical variables.
    """
    try:
        # 1. Identify problematic categorical columns
        cat_cols = df.select_dtypes(include=['object', 'category']).columns
        problematic_cols = []

        for col in cat_cols:
            unique_vals = df[col].dropna().unique()
            if len(unique_vals) <= 1:
                problematic_cols.append(col)

        # 2. Handle problematic columns
        if problematic_cols:
            print(f"Note: Columns {problematic_cols} have ≤1 class. Using mode imputation.")
            for col in problematic_cols:
                mode_val = df[col].mode()[0] if not df[col].mode().empty else np.nan
                df[col] = df[col].fillna(mode_val)

        # 3. Separate numeric and remaining categorical columns
        numeric_cols = df.select_dtypes(include=np.number).columns
        remaining_cat_cols = [col for col in cat_cols if col not in problematic_cols]

        # 4. Create appropriate imputers
        imputers = {}

        # Numeric imputer
        if len(numeric_cols) > 0:
            imputers['numeric'] = IterativeImputer(
                estimator=RandomForestRegressor(n_estimators=100),
                max_iter=max_iter,
                random_state=random_state
            )

        # Categorical imputer (only if we have multi-class columns)
        if len(remaining_cat_cols) > 0:
            imputers['categorical'] = IterativeImputer(
                estimator=RandomForestClassifier(n_estimators=100),
                max_iter=max_iter,
                random_state=random_state,
                initial_strategy='most_frequent'
            )

        # 5. Perform imputation
        if len(imputers) == 0:
            return df  # No columns left to impute

        # Numeric imputation
        if 'numeric' in imputers:
            df[numeric_cols] = imputers['numeric'].fit_transform(df[numeric_cols])

        # Categorical imputation
        if 'categorical' in imputers:
            # Need to encode categories first
            from sklearn.preprocessing import OrdinalEncoder
            encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
            encoded = encoder.fit_transform(df[remaining_cat_cols])
            imputed = imputers['categorical'].fit_transform(encoded)
            df[remaining_cat_cols] = encoder.inverse_transform(imputed)

        return df

    except Exception as e:
        print(f"MICE imputation failed: {str(e)}")
        print("Falling back to simple imputation...")
        return df.fillna(df.mean())  # Numeric fallback


def impute_mice_forest_robust(df_missing, n_iterations=5):
    """Robust MICE Forest implementation with comprehensive error handling"""
    try:
        # Convert data to pandas DataFrame if it isn't already
        df_missing = pd.DataFrame(df_missing.copy())

        # Convert object columns to category for efficiency
        cat_cols = df_missing.select_dtypes(include=['object']).columns
        for col in cat_cols:
            df_missing[col] = df_missing[col].astype('category')

        # Initialize kernel with current API
        kernel = mf.ImputationKernel(
            data=df_missing,
            random_state=42,
         )

        # Run MICE algorithm with progress tracking
        kernel.mice(n_iterations, verbose=True)

        # Return the first imputed dataset with validation
        imputed = kernel.complete_data(dataset=0)

        # Ensure no NaN values remain
        if imputed.isna().any().any():
            print("Warning: MICE produced NaN values - filling with mode")
            imputed = imputed.fillna(imputed.mode().iloc[0])
        return imputed

    except Exception as e:
        print(f"Error in MICE Forest imputation: {str(e)}")
        print("Falling back to mode imputation")
        return impute_mode(df_missing)

def impute_mice_forest(df_missing, n_iterations=5):
    try:
        kernel = mf.ImputationKernel(data=df_missing.astype('category'), random_state=42)
        kernel.mice(n_iterations, verbose=False)
        return kernel.complete_data(dataset=0)
    except Exception as e:
        print(f"Error in MICE Forest imputation: {str(e)}")
        # Consider falling back to a simpler method
        return impute_mode(df_missing)  # Fallback to mode imputation


def impute_missforest(df_missing, max_iter=10):
    """MissForest imputation using Random Forests"""

    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

##4    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
    encoded = encoder.fit_transform(df_missing)
    imputer = IterativeImputer(estimator=RandomForestRegressor(n_estimators=100, random_state=42),
                              max_iter=max_iter, random_state=42)
    imputed_encoded = imputer.fit_transform(encoded)

    return pd.DataFrame(encoder.inverse_transform(imputed_encoded), columns=df_missing.columns)

def impute_softimpute(df_missing):
    """Matrix completion by iterative soft thresholding of SVD decompositions"""

    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

##5    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
    encoded = encoder.fit_transform(df_missing)
    imputed = SoftImpute().fit_transform(encoded)
    return pd.DataFrame(encoder.inverse_transform(imputed), columns=df_missing.columns)

def impute_nuclear_norm(df_missing):
    """Matrix completion via convex optimization of nuclear norm"""
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

##11    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
    encoded = encoder.fit_transform(df_missing)
    imputed = NuclearNormMinimization().fit_transform(encoded)
    return pd.DataFrame(encoder.inverse_transform(imputed), columns=df_missing.columns)
```

```python

def impute_autoencoder(df_missing, epochs=50, batch_size=32, verbose=0):

    # 1. Data Encoding - Process in chunks if large
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

    if len(df_missing) > 100000:  # For very large datasets
        # Process in chunks to avoid memory spikes
        chunks = []
        for i in range(0, len(df_missing), 50000):
            chunk = df_missing.iloc[i:i+50000]
            chunks.append(encoder.fit_transform(chunk))
        encoded = np.concatenate(chunks)
    else:
        encoded = encoder.fit_transform(df_missing)

    # 2. Normalization - track min/max per column for reconstruction
    col_mins = np.nanmin(encoded, axis=0)
    col_maxs = np.nanmax(encoded, axis=0)

    # Handle columns with all NaN
    valid_cols = ~np.isnan(col_mins)
    if not np.all(valid_cols):
        # Fill completely missing columns with mode first
        for col in np.where(~valid_cols)[0]:
            mode = stats.mode(encoded[:, valid_cols], axis=0, keepdims=True).mode
            encoded[:, col] = mode[0, 0] if mode.size > 0 else 0

    # Normalize
    X = (encoded - col_mins) / (col_maxs - col_mins + 1e-10)
    X = np.nan_to_num(X, nan=0)
    mask = (~np.isnan(encoded)).astype(float)

    # 3. Build Memory-Efficient Autoencoder
    input_dim = X.shape[1]
    encoding_dim = min(16, max(2, input_dim // 8))  # Reduced capacity

    # Use functional API with explicit layer cleanup
    input_layer = Input(shape=(input_dim,))
    x = Dense(64, activation='relu')(input_layer)
    x = Dense(encoding_dim, activation='relu')(x)
    x = Dense(64, activation='relu')(x)
    output_layer = Dense(input_dim, activation='sigmoid')(x)

    autoencoder = Model(inputs=input_layer, outputs=output_layer)
    autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

    # 4. Training with Batch Processing
    if len(X) > batch_size * 10:  # Large dataset - use generator
        def data_generator(X_data, batch_size):
            while True:
                idx = np.random.randint(0, len(X_data), batch_size)
                yield X_data[idx], X_data[idx]

        autoencoder.fit(
            data_generator(X, batch_size),
            steps_per_epoch=len(X)//batch_size,
            epochs=epochs,
            verbose=verbose
        )
    else:  # Small dataset
        autoencoder.fit(
            X, X,
            epochs=epochs,
            batch_size=batch_size,
            verbose=verbose
        )

    # 5. Imputation with Batch Prediction
    X_imputed = np.zeros_like(X)
    for i in range(0, len(X), batch_size):
        batch = X[i:i+batch_size]
        X_imputed[i:i+batch_size] = autoencoder.predict(batch, verbose=0)

    # Only replace missing values
    X_imputed = mask * X + (1 - mask) * X_imputed

    # 6. Denormalize and Post-process
    X_imputed = X_imputed * (col_maxs - col_mins + 1e-10) + col_mins
    X_imputed = np.round(X_imputed).astype(int)

    # Clip to valid category ranges
    for i in range(X_imputed.shape[1]):
        X_imputed[:, i] = np.clip(
            X_imputed[:, i],
            0,
            len(encoder.categories_[i])-1 if i < len(encoder.categories_) else 0
        )

    # 7. Clean up TensorFlow session and large variables
    K.clear_session()
    del autoencoder, input_layer, output_layer, x
    gc.collect()

    # 8. Return imputed data
    return pd.DataFrame(encoder.inverse_transform(X_imputed), columns=df_missing.columns)

def impute_gbm_classifier(df_missing):
    """Gradient Boosted Trees classifier for each column with missing values"""
    imputed = df_missing.copy()
    for col in df_missing.columns:
        mask = df_missing[col].isna()
        if mask.sum() == 0:
            continue

        # Prepare data
        X_train = df_missing[~mask].drop(columns=[col])
        y_train = df_missing.loc[~mask, col]
        X_test = df_missing[mask].drop(columns=[col])

        # Encode categorical data.
        encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

##6        encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
        X_train_enc = encoder.fit_transform(X_train)
        X_test_enc = encoder.transform(X_test)

        # Train classifier
        clf = HistGradientBoostingClassifier(random_state=42)
        clf.fit(X_train_enc, y_train)

        # Predict missing values
        imputed.loc[mask, col] = clf.predict(X_test_enc)
    return imputed
```

```python
def impute_random_forest(df_missing, n_estimators=100, max_depth=None, random_state=42):
    """
    Random Forest imputation for missing values using sklearn's RandomForestRegressor/RandomForestClassifier.
    Automatically detects whether each column is numeric or categorical and uses appropriate model.
    """
    df_imputed = df_missing.copy()

    for col in df_imputed.columns:
        # Check if column has missing values
        if df_imputed[col].isna().sum() > 0:
            # Split data into observed and missing
            observed = df_imputed[~df_imputed[col].isna()]
            missing = df_imputed[df_imputed[col].isna()]

            # Skip if no observed data
            if len(observed) == 0:
                continue

            # Prepare features (all columns except current) and target (current column)
            X_train = observed.drop(columns=[col])
            y_train = observed[col]
            X_test = missing.drop(columns=[col])

            # Encode categorical features if needed
            categorical_cols = X_train.select_dtypes(include=['object', 'category']).columns
            if len(categorical_cols) > 0:
                encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
                X_train[categorical_cols] = encoder.fit_transform(X_train[categorical_cols])
                X_test[categorical_cols] = encoder.transform(X_test[categorical_cols])

            # Determine if target is numeric or categorical
            if pd.api.types.is_numeric_dtype(y_train):
                # Use RandomForestRegressor for numeric targets
                model = RandomForestRegressor(
                    n_estimators=n_estimators,
                    max_depth=max_depth,
                    random_state=random_state
                )
                model.fit(X_train, y_train)
                preds = model.predict(X_test)
            else:
                # Use RandomForestClassifier for categorical targets
                model = RandomForestClassifier(
                    n_estimators=n_estimators,
                    max_depth=max_depth,
                    random_state=random_state
                )
                model.fit(X_train, y_train)
                preds = model.predict(X_test)

            # Fill missing values
            df_imputed.loc[df_imputed[col].isna(), col] = preds
    return df_imputed
```

```python
def impute_truncated_normal(df_missing, std_dev_factor=0.5):
    """
    Imputes missing values from a normal distribution truncated to observed min/max.
    Preserves overall distribution while adding realistic variance.

    """
    from scipy.stats import truncnorm

    df_imputed = df_missing.copy()

    for col in df_imputed.select_dtypes(include=['number']):
        missing_mask = df_imputed[col].isna()
        observed = df_imputed[col].dropna()

        if len(observed) > 1:
            mean = observed.mean()
            std = observed.std() * std_dev_factor
            a = (observed.min() - mean) / std
            b = (observed.max() - mean) / std

            # Generate truncated normal samples
            samples = truncnorm.rvs(a, b, loc=mean, scale=std,
                                  size=missing_mask.sum())
            df_imputed.loc[missing_mask, col] = samples

    return df_imputed


def impute_quantile_matching(df_missing, n_quantiles=10):
    """
    Matches the quantile position of missing values to randomly sampled observed values.
    Preserves the original distribution exactly.
    """
    df_imputed = df_missing.copy()

    for col in df_imputed.select_dtypes(include=['number']):
        missing_mask = df_imputed[col].isna()
        observed = df_imputed[col].dropna().sort_values()

        if len(observed) > n_quantiles:
            # Create quantile bins
            quantiles = np.linspace(0, 1, n_quantiles + 1)
            bin_edges = observed.quantile(quantiles)

            # Randomly assign missing values to quantiles
            random_quantiles = np.random.choice(quantiles[:-1], size=missing_mask.sum())

            # Sample uniformly within each quantile bin
            imputed_values = []
            for q in random_quantiles:
                lower = bin_edges[q]
                upper = bin_edges[q + 1]
                sample = np.random.uniform(lower, upper)
                imputed_values.append(sample)

            df_imputed.loc[missing_mask, col] = imputed_values

    return df_imputed

def impute_robust_scaled_noise(df_missing, noise_scale=0.1):
    """
    Scales observed values to median/IQR, adds noise, then transforms back.
    Resistant to outliers in the imputation process.
    """
    from sklearn.preprocessing import RobustScaler

    df_imputed = df_missing.copy()

    for col in df_imputed.select_dtypes(include=['number']):
        missing_mask = df_imputed[col].isna()
        observed = df_imputed[col].dropna()

        if len(observed) == 0:
            # No observed data; fallback imputation
            df_imputed.loc[missing_mask, col] = 0  # or another global default
            continue

        # Robust scaling
        scaler = RobustScaler()
        scaled = scaler.fit_transform(observed.values.reshape(-1, 1)).flatten()

        # Impute with noise
        noise = np.random.normal(loc=0, scale=noise_scale, size=missing_mask.sum())
        imputed_scaled = np.random.choice(scaled, size=missing_mask.sum()) + noise

        # Inverse transform
        imputed_values = scaler.inverse_transform(imputed_scaled.reshape(-1, 1)).flatten()
        df_imputed.loc[missing_mask, col] = imputed_values

    return df_imputed


def impute_local_outlier_aware(df_missing, k=5, contamination=0.1):
    """
    Uses local neighborhood information while downweighting outliers.

    """
    from sklearn.neighbors import LocalOutlierFactor
    from sklearn.impute import KNNImputer

    df_imputed = df_missing.copy()

    for col in df_imputed.select_dtypes(include=['number']):
        missing_mask = df_imputed[col].isna()

        if missing_mask.any():
            # Detect outliers
            lof = LocalOutlierFactor(n_neighbors=k, contamination=contamination)
            outlier_scores = lof.fit_predict(df_imputed.select_dtypes(include=['number']))
            weights = np.where(outlier_scores == -1, 0.1, 1)  # Downweight outliers

            # Weighted KNN imputation
            imputer = KNNImputer(n_neighbors=k, weights='distance')
            imputed_values = imputer.fit_transform(
                df_imputed.select_dtypes(include=['number']),
                sample_weight=weights
            )[:, df_imputed.columns.get_loc(col)]

            df_imputed.loc[missing_mask, col] = imputed_values[missing_mask]

    return df_imputed


def impute_spline_with_uncertainty(df_missing, order=3, n_samples=100):
    """
    Fits smoothing splines to observed data and samples from prediction interval.
    Ideal for time-series or ordered data with measurement error.
    """
    df_imputed = df_missing.copy()

    for col in df_imputed.select_dtypes(include=['number']):
        missing_mask = df_imputed[col].isna()
        observed = df_imputed[col].dropna().reset_index(drop=True)  # reset index
        x_obs = np.arange(len(observed))  # x positions now match

        if len(observed) > order + 1:
            imputed_samples = []
            x_missing = np.where(missing_mask)[0]

            for _ in range(n_samples):
                x_resampled, y_resampled = resample(x_obs, observed.values)

                # Sort by x to satisfy spline requirement
                sorted_indices = np.argsort(x_resampled)
                x_resampled = x_resampled[sorted_indices]
                y_resampled = y_resampled[sorted_indices]

                spline = UnivariateSpline(x_resampled, y_resampled, k=order)

                # Interpolate missing values based on their relative position
                imputed_samples.append(spline(x_missing))

            # Assign median of sampled imputations
            df_imputed.loc[missing_mask, col] = np.median(imputed_samples, axis=0)

    return df_imputed


def impute_copula2(df_missing, n_iter=100):
    """
    Impute missing values using a Gaussian copula. If data is too sparse, falls back to mean imputation.
    """
    import pandas as pd
    import numpy as np
    from copulas.multivariate import GaussianMultivariate

    df_imputed = df_missing.copy()
    numeric_cols = df_imputed.select_dtypes(include=['number']).columns

    complete_data = df_imputed[numeric_cols].dropna()

    # Fallback: use mean imputation if no complete rows
    if complete_data.empty:
        print("⚠️ No complete rows found. Falling back to mean imputation.")
        for col in numeric_cols:
            mean_val = df_imputed[col].mean()
            df_imputed[col].fillna(mean_val, inplace=True)
        return df_imputed

    # Fit copula model
    copula = GaussianMultivariate()
    copula.fit(complete_data)

    # Gibbs-style conditional sampling
    for _ in range(n_iter):
        for col in numeric_cols:
            missing_mask = df_imputed[col].isna()
            if missing_mask.any():
                cond_cols = [c for c in numeric_cols if c != col]
                condition_data = df_imputed.loc[missing_mask, cond_cols]

                # Only use rows with complete conditioning values
                valid_mask = ~condition_data.isna().any(axis=1)
                condition_data = condition_data[valid_mask]

                if not condition_data.empty:
                    try:
                        samples = copula.conditional_distribution(cond_cols, condition_data).sample()
                        df_imputed.loc[condition_data.index, col] = samples[col].values
                    except Exception as e:
                        print(f"⚠️ Copula failed for '{col}'. Using mean as fallback. Error: {e}")
                        mean_val = df_imputed[col].mean()
                        df_imputed.loc[missing_mask, col] = mean_val
                else:
                    print(f"⚠️ Not enough data to conditionally impute '{col}'. Using mean.")
                    mean_val = df_imputed[col].mean()
                    df_imputed.loc[missing_mask, col] = mean_val

    return df_imputed

def impute_copula2_(df_missing, n_iter=100):
    """
    Models multivariate dependence structure using copulas before imputation.
    Preserves complex correlation structures better than MVN.

    """
    from copulas.multivariate import GaussianMultivariate

    df_imputed = df_missing.copy()
    numeric_cols = df_imputed.select_dtypes(include=['number']).columns

    # Fit copula model
    copula = GaussianMultivariate()
    copula.fit(df_imputed[numeric_cols].dropna())

    # Gibbs sampling imputation
    for _ in range(n_iter):
        for col in numeric_cols:
            missing_mask = df_imputed[col].isna()
            if missing_mask.any():
                # Sample from conditional distribution
                cond_cols = [c for c in numeric_cols if c != col]
                samples = copula.conditional_sample(
                    df_imputed[cond_cols].values[missing_mask],
                    [col]
                )
                df_imputed.loc[missing_mask, col] = samples[col].values

    return df_imputed


def impute_optimal_transport2(df_missing, reg=0.1):
    """
    Uses optimal transport theory to match distributions before/after imputation.
    Minimizes Wasserstein distance between observed and imputed distributions.

    """

    df_imputed = df_missing.copy()

    for col in df_imputed.select_dtypes(include=['number']):
        missing_mask = df_imputed[col].isna()
        observed = df_imputed[col].dropna()

        if len(observed) > 1:
            # Create source (observed) and target (uniform) distributions
            source = observed.values.reshape(-1, 1)
            target = np.linspace(observed.min(), observed.max(),
                               len(source)).reshape(-1, 1)

            # Compute cost matrix
            M = np.abs(source - target.T) ** 2

            # Solve transport problem
            if len(source) < 100:  # Exact OT for small problems
                transport = emd(np.ones(len(source))/len(source),
                              np.ones(len(target))/len(target),
                              M)
            else:  # Approximate OT for larger problems
                transport = sinkhorn(np.ones(len(source))/len(source),
                                   np.ones(len(target))/len(target),
                                   M, reg=reg)

            # Sample from transported distribution
            probs = transport.sum(axis=1)
            probs = probs / probs.sum()
            samples = np.random.choice(target.flatten(),
                                     size=missing_mask.sum(),
                                     p=probs)
            df_imputed.loc[missing_mask, col] = samples

    return df_imputed
```

```python
def impute_conditional_probability(df_missing, condition_cols=None):
    df_imputed = df_missing.copy()

    for col in df_imputed.select_dtypes(include=['object', 'category']):
        missing_mask = df_imputed[col].isna()

        if condition_cols is None:
            condition_cols = [c for c in df_imputed.columns if c != col]

        # Conditional probabilities
        prob_table = (df_imputed[~missing_mask]
                      .groupby(condition_cols)[col]
                      .value_counts(normalize=True)
                      .unstack())

        for idx in df_imputed[missing_mask].index:
            conditions = df_imputed.loc[idx, condition_cols]

            if conditions.isnull().any():
                # fallback to marginal
                marginal = df_imputed[col].value_counts(normalize=True)
                if not marginal.empty:
                    df_imputed.loc[idx, col] = np.random.choice(marginal.index, p=marginal.values)
                else:
                    categories = df_imputed[col].dropna().unique()
                    if len(categories) > 0:
                        df_imputed.loc[idx, col] = np.random.choice(categories)
                    else:
                        df_imputed.loc[idx, col] = 'missing'
                continue

            try:
                probs = prob_table.loc[tuple(conditions)].values
                categories = prob_table.columns
                df_imputed.loc[idx, col] = np.random.choice(categories, p=probs)
            except:
                # fallback to marginal
                marginal = df_imputed[col].value_counts(normalize=True)
                if not marginal.empty:
                    df_imputed.loc[idx, col] = np.random.choice(marginal.index, p=marginal.values)
                else:
                    categories = df_imputed[col].dropna().unique()
                    if len(categories) > 0:
                        df_imputed.loc[idx, col] = np.random.choice(categories)
                    else:
                        df_imputed.loc[idx, col] = 'missing'

    return df_imputed

def impute_association_rules(df_missing, min_support=0.1, min_confidence=0.5):
    """
    Uses frequent itemset mining to find association rules for imputation.
    Use patterns and associations within the data (like those found with the Apriori algorithm) to impute likely values.

    """
    from mlxtend.frequent_patterns import apriori, association_rules

    df_imputed = df_missing.copy()
    df_encoded = pd.get_dummies(df_imputed.astype(str))

    # Find frequent itemsets
    frequent_itemsets = apriori(df_encoded, min_support=min_support, use_colnames=True)

    # Generate rules
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)

    # Sort rules by confidence and lift
    rules = rules.sort_values(['confidence', 'lift'], ascending=False)

    # Apply rules for imputation
    for col in df_imputed.select_dtypes(include=['object', 'category']):
        missing_mask = df_imputed[col].isna()
        col_dummy = f"{col}_{{}}"

        for idx in df_imputed[missing_mask].index:
            # Find applicable rules
            for _, rule in rules.iterrows():
                antecedents = set(rule['antecedents'])
                consequents = set(rule['consequents'])

                # Check if rule applies
                row_values = set(df_encoded.columns[df_encoded.loc[idx] == 1])
                if antecedents.issubset(row_values):
                    # Find matching consequent for our column
                    for consequent in consequents:
                        if consequent.startswith(col + "_"):
                            df_imputed.loc[idx, col] = consequent.split('_')[1]
                            break
                    break

    return df_imputed


def impute_kmodes(df_missing, n_clusters=3, max_iter=5):
    """
    Uses k-modes clustering to impute missing categorical values.

    """
    from kmodes.kmodes import KModes

    df_imputed = df_missing.copy()
    df_encoded = df_imputed.apply(lambda x: x.factorize()[0] if x.dtype == 'object' else x)

    # Fit k-modes
    km = KModes(n_clusters=n_clusters, max_iter=max_iter)
    clusters = km.fit_predict(df_encoded.fillna(-1))

    # Impute using cluster modes
    for col in df_imputed.select_dtypes(include=['object', 'category']):
        missing_mask = df_imputed[col].isna()

        for cluster in range(n_clusters):
            cluster_mask = (clusters == cluster) & (~missing_mask)
            if sum(cluster_mask) > 0:
                mode = df_imputed.loc[cluster_mask, col].mode()
                if len(mode) > 0:
                    df_imputed.loc[(clusters == cluster) & missing_mask, col] = mode[0]

    return df_imputed


from pgmpy.estimators import PC
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.sampling import BayesianModelSampling

def impute_bayesian_network(df_missing, n_samples=1000):
    """
    Uses Bayesian network structure learning to model dependencies.
    Impute based on posterior predictive distributions.
    """
    df_imputed = df_missing.copy()
    df_encoded = df_imputed.apply(lambda x: x.factorize()[0] if x.dtype == 'object' else x)

    # Learn structure using PC algorithm
    est = PC(df_encoded.dropna())
    model = est.estimate(variant="stable", max_cond_vars=4)

    # Fit parameters and sample
    bn = DiscreteBayesianNetwork(model.edges())  # Use DiscreteBayesianNetwork
    bn.fit(df_encoded.dropna())
    sampler = BayesianModelSampling(bn)
    samples = sampler.forward_sample(size=n_samples)

    # Impute by matching similar complete cases
    for col in df_imputed.columns:
        missing_mask = df_imputed[col].isna()
        if missing_mask.any():
            # Find most similar samples
            known_cols = [c for c in df_imputed.columns if c != col and not df_imputed[c].isna().all()]
            if known_cols:
                impute_values = []
                for idx in df_imputed[missing_mask].index:
                    row = df_encoded.loc[idx, known_cols]
                    distances = (samples[known_cols] - row).abs().sum(axis=1)
                    closest = distances.idxmin()
                    impute_values.append(samples.loc[closest, col])

                df_imputed.loc[missing_mask, col] = impute_values

    return df_imputed


def impute_fuzzy_matching(df_missing, threshold=0.8):
    """
    Uses fuzzy string matching to find similar rows for imputation.

    """
    df_imputed = df_missing.copy()
    str_cols = df_imputed.select_dtypes(include=['object']).columns
    num_cols = df_imputed.select_dtypes(include=[np.number]).columns

    for col in str_cols:
        missing_mask = df_imputed[col].isna()
        if not missing_mask.any():
            continue

        # Create string representations for matching
        df_imputed['_match_str'] = df_imputed[str_cols].fillna('').astype(str).agg(' '.join, axis=1)

        for idx in df_imputed[missing_mask].index:
            target_str = df_imputed.loc[idx, '_match_str']

            # Find best match among complete cases
            best_match = None
            best_score = 0

            for other_idx in df_imputed[~missing_mask].index:
                other_str = df_imputed.loc[other_idx, '_match_str']
                score = fuzz.token_set_ratio(target_str, other_str)/100

                if score > best_score and score >= threshold:
                    best_score = score
                    best_match = other_idx

            if best_match is not None:
                df_imputed.loc[idx, col] = df_imputed.loc[best_match, col]

        df_imputed.drop('_match_str', axis=1, inplace=True)

    ### Numeric: Mean Imputation
    for col in num_cols:
        mean_val = df_imputed[col].mean(skipna=True)
        df_imputed[col].fillna(mean_val, inplace=True)


    return df_imputed
```

```python
#1. Diffusion Model Imputation (Continuous Diffusion)

def impute_diffusion(df_missing, steps=100, epochs=20):
    """Diffusion model for categorical imputation"""
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Normalize
    min_val = np.nanmin(encoded)
    max_val = np.nanmax(encoded)
    X = (encoded - min_val) / (max_val - min_val)
    X = np.nan_to_num(X, nan=0)
    mask = (~np.isnan(encoded)).astype(float)

    # Diffusion parameters
    beta = np.linspace(1e-4, 0.1, steps)
    alpha = 1 - beta
    alpha_bar = np.cumprod(alpha)

    # UNet model
    def build_unet():
        inputs = Input(shape=(X.shape[1],))
        x = Dense(128, activation='relu')(inputs)
        x = Dense(128, activation='relu')(x)
        outputs = Dense(X.shape[1])(x)
        return Model(inputs, outputs)

    model = build_unet()
    model.compile(optimizer='adam', loss='mse')

    # Training
    for epoch in range(epochs):
        for _ in range(len(X)):
            t = np.random.randint(1, steps)
            x0 = X[np.random.choice(len(X))]
            eps = np.random.normal(size=x0.shape)

            xt = np.sqrt(alpha_bar[t]) * x0 + np.sqrt(1-alpha_bar[t]) * eps
            loss = model.train_on_batch(xt.reshape(1,-1), eps.reshape(1,-1))

    # Imputation
    xt = np.random.normal(size=X.shape)
    for t in reversed(range(steps)):
        eps_pred = model.predict(xt, verbose=0)
        xt = (xt - (1-alpha[t])/np.sqrt(1-alpha_bar[t]) * eps_pred) / np.sqrt(alpha[t])
        if t > 0:
            z = np.random.normal(size=X.shape)
            xt += np.sqrt(beta[t]) * z

    # Only replace missing values
    X_imputed = mask * X + (1 - mask) * xt

    # Convert back
    X_imputed = X_imputed * (max_val - min_val) + min_val
    X_imputed = np.round(X_imputed).astype(int)
    X_imputed = np.clip(X_imputed, 0, len(encoder.categories_[0])-1)


    return pd.DataFrame(encoder.inverse_transform(X_imputed), columns=df_missing.columns)

#3. Neural Process Imputation
def impute_neural_process(df_missing, epochs=50):
    """Neural Process for few-shot imputation"""
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Normalize
    min_val = np.nanmin(encoded, axis=0)
    max_val = np.nanmax(encoded, axis=0)

    # Use a robust range for normalization, handling constant/NaN columns
    range_vals = max_val - min_val
    range_vals[range_vals == 0] = 1e-8 # Avoid division by zero
    min_val[np.isnan(min_val)] = 0 # Handle all NaN columns in min/max calculation


    X = (encoded - min_val) / range_vals
    X = np.nan_to_num(X, nan=0)
    mask = (~np.isnan(encoded)).astype(float)

    # Neural Process components
    def build_encoder():
        inputs = Input(shape=(X.shape[1],))
        x = Dense(64, activation='relu')(inputs)
        x = Dense(64, activation='relu')(x)
        return Model(inputs, x)


    def build_decoder(latent_input_dim, output_dim):
        # The input to the decoder is the concatenated latent vector z
        # z_context is 64, z_target is 64, so z is 64 + 64 = 128
        # Therefore, latent_input_dim should be 128
        latent_inputs = Input(shape=(latent_input_dim,))
        x = Dense(64, activation='relu')(latent_inputs) # First layer processing the combined latent vector
        outputs = Dense(output_dim, activation='sigmoid')(x) # Output dimension matches original data features
        return Model(latent_inputs, outputs)


    def build_decoder_():
        latent_inputs = Input(shape=(64,))
        x = Dense(64, activation='relu')(latent_inputs)
        outputs = Dense(X.shape[1], activation='sigmoid')(x)
        return Model(latent_inputs, outputs)

    encoder_model = build_encoder()
    #decoder_model = build_decoder()
    decoder_input_dim = 64 + 64 # Output dim of encoder_model for context + Output dim of encoder_model for target
    decoder_model = build_decoder(decoder_input_dim, X.shape[1])

    # Combined model
    inputs = Input(shape=(X.shape[1],))
    context_mask = Input(shape=(X.shape[1],))

    # Get latent representation
    z_context = encoder_model(inputs * context_mask)
    z_target = encoder_model(inputs)

    # Combine
    z = Concatenate()([z_context, z_target])
    outputs = decoder_model(z)

    model = Model(inputs=[inputs, context_mask], outputs=outputs)
    model.compile(optimizer='adam', loss='mse')


    # Training
    for epoch in range(epochs):
        # It's generally better to train on shuffled batches rather than one sample at a time
        indices = np.arange(len(X))
        np.random.shuffle(indices)
        for i in indices:
             # Generate a random context mask for this instance
            context_mask_instance = np.random.binomial(1, 0.5, size=X.shape[1])

            model.train_on_batch([X[i:i+1], context_mask_instance.reshape(1,-1)], X[i:i+1])

    # Imputation
    X_imputed = X.copy()
    for i in range(len(X)):
        missing = mask[i] == 0
        if missing.any():
            context = mask[i].copy()
            pred = model.predict([X[i:i+1], context_mask_instance.reshape(1,-1)], verbose=0)
            X_imputed[i,missing] = pred[0,missing]

    # Convert back
    imputed_encoded = X_imputed * range_vals + min_val
    imputed_encoded = np.round(imputed_encoded).astype(int)



    # Clip to valid category ranges
    # This is crucial after rounding for ordinal/categorical data
    for i in range(imputed_encoded.shape[1]):
         if i < len(encoder.categories_): # Check if it's a categorical column that was encoded
             valid_categories = encoder.categories_[i]
             min_cat_val = 0 # OrdinalEncoder starts from 0
             max_cat_val = len(valid_categories) - 1
             imputed_encoded[:, i] = np.clip(imputed_encoded[:, i], min_cat_val, max_cat_val)
         else:
             # For numeric columns that might have been passed through the encoder (less common scenario)
             # You might need a different clipping strategy or skip clipping
             pass # Or clip based on original min/max if known


    return pd.DataFrame(encoder.inverse_transform(imputed_encoded), columns=df_missing.columns)
```

```python
def impute_xgboost_lss(df_missing):
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    imputed = encoded.copy()
    for col in range(encoded.shape[1]):
        mask = ~np.isnan(encoded[:, col])
        if np.sum(mask) == 0:
            continue  # skip if entire column is missing

        # Features without the current column
        features = [i for i in range(encoded.shape[1]) if i != col]

        # Train mean model
        X_train = encoded[mask][:, features]
        y_train = encoded[mask, col]
        model_mean = xgb.XGBRegressor(objective='reg:squarederror')
        model_mean.fit(X_train, y_train)

        # Train variance model
        residuals = (y_train - model_mean.predict(X_train)) ** 2
        model_var = xgb.XGBRegressor(objective='reg:squarederror')
        model_var.fit(X_train, residuals)

        # Impute missing values
        missing_mask = ~mask
        X_missing = encoded[missing_mask][:, features]
        if len(X_missing) > 0:
            pred_mean = model_mean.predict(X_missing)
            pred_var = model_var.predict(X_missing)

            # Avoid negative/zero variance
            pred_var = np.maximum(pred_var, 1e-6)

            imputed_values = np.random.normal(pred_mean, np.sqrt(pred_var))
            imputed[missing_mask, col] = imputed_values

    # Round and convert back to original categories
    imputed = np.round(imputed).astype(int)
    for col, cats in enumerate(encoder.categories_):
        imputed[:, col] = np.clip(imputed[:, col], 0, len(cats) - 1)

    df_imputed = pd.DataFrame(encoder.inverse_transform(imputed), columns=df_missing.columns)
    return df_imputed
```

```python
def impute_mem_transformer(df_missing, d_model=64, num_heads=4, epochs=50):
    """Transformer with memory-efficient attention for large datasets
    Transformer with Memory Compressed Attention
    """
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Normalize and prepare masks
    min_val, max_val = np.nanmin(encoded), np.nanmax(encoded)
    X = (encoded - min_val) / (max_val - min_val)
    X = np.nan_to_num(X, nan=0)
    mask = (~np.isnan(encoded)).astype(float)

    # Build model with memory compressed attention
    inputs = Input(shape=(X.shape[1],))
    mask_input = Input(shape=(X.shape[1],))

    # Memory compression layer
    compressed = Dense(d_model//2)(inputs)

    # Multi-head attention with compression
    attn_output = MultiHeadAttention(
        num_heads=num_heads,
        key_dim=d_model//num_heads,
        attention_axes=1
    )(compressed, compressed)

    # Output
    outputs = Dense(X.shape[1], activation='sigmoid')(attn_output)

    model = Model(inputs=[inputs, mask_input], outputs=outputs)
    model.compile(optimizer='adam', loss='mse')

    # Train with masked loss
    model.fit([X, mask], X, epochs=epochs, batch_size=32, verbose=0)

    # Impute missing values
    X_hat = model.predict([X, mask], verbose=0)
    X_imputed = mask * X + (1 - mask) * X_hat

    # Post-processing
    X_imputed = X_imputed * (max_val - min_val) + min_val
    X_imputed = np.round(X_imputed).astype(int)
    X_imputed = np.clip(X_imputed, 0, len(encoder.categories_[0])-1)
    return pd.DataFrame(encoder.inverse_transform(X_imputed), columns=df_missing.columns)


def impute_neural_ode(df_missing, epochs=50):
    """Neural ODE for continuous-depth imputation
    Neural Differential Equations
"""
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Normalize
    min_val, max_val = np.nanmin(encoded), np.nanmax(encoded)
    X = (encoded - min_val) / (max_val - min_val)
    X = np.nan_to_num(X, nan=0)
    mask = (~np.isnan(encoded)).astype(float)

    # Define ODE function
    class ODEFunc(nn.Module):
        def __init__(self, dim):
            super().__init__()
            self.net = nn.Sequential(
                nn.Linear(dim, 64),
                nn.ReLU(),
                nn.Linear(64, dim)
            )

        def forward(self, t, x):
            return self.net(x)

    # Convert to PyTorch tensors
    X_t = torch.FloatTensor(X)
    mask_t = torch.FloatTensor(mask)

    # Model
    odefunc = ODEFunc(X.shape[1])
    optimizer = torch.optim.Adam(odefunc.parameters(), lr=0.01)

    # Training
    for epoch in range(epochs):
        optimizer.zero_grad()

        # Solve ODE
        t = torch.linspace(0, 1, 2)
        pred = odeint(odefunc, X_t, t, method='dopri5')[-1]

        # Masked loss
        loss = torch.mean((pred - X_t)**2 * mask_t)
        loss.backward()
        optimizer.step()

    # Imputation
    with torch.no_grad():
        X_hat = odeint(odefunc, X_t, t, method='dopri5')[-1].numpy()

    X_imputed = mask * X + (1 - mask) * X_hat

    # Post-processing
    X_imputed = X_imputed * (max_val - min_val) + min_val
    X_imputed = np.round(X_imputed).astype(int)
    X_imputed = np.clip(X_imputed, 0, len(encoder.categories_[0])-1)

    return pd.DataFrame(encoder.inverse_transform(X_imputed), columns=df_missing.columns)



def impute_hypernetwork(df_missing, inner_epochs=5, outer_epochs=20):
    """Hypernetwork that generates imputation parameters
    Hypernetwork Imputation
    """
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Normalize
    min_val, max_val = np.nanmin(encoded), np.nanmax(encoded)
    X = (encoded - min_val) / (max_val - min_val)
    X = np.nan_to_num(X, nan=0)
    mask = (~np.isnan(encoded)).astype(float)

    # Hypernetwork architecture
    def build_hypernetwork():
        inp = Input(shape=(X.shape[1],))
        x = Dense(128, activation='relu')(inp)
        x = Dense(256, activation='relu')(x)

        # Output weights for target network
        w1 = Dense(64 * 64)(x)
        b1 = Dense(64)(x)
        w2 = Dense(64 * X.shape[1])(x)
        b2 = Dense(X.shape[1])(x)

        return Model(inp, [w1, b1, w2, b2])

    # Target network using Lambda layers to do matmul and activations
    def build_final_model(hypernet):
        x_in = Input(shape=(X.shape[1],))
        w1, b1, w2, b2 = hypernet(x_in)

        w1 = Reshape((X.shape[1], 64))(w1)  # Corrected shape to match input dims
        w2 = Reshape((64, X.shape[1]))(w2)
        b1 = Reshape((1, 64))(b1)
        b2 = Reshape((1, X.shape[1]))(b2)

        # Compute h = relu(x @ w1 + b1)
        def compute_hidden(inputs):
            x, w1, b1 = inputs
            x_exp = tf.expand_dims(x, 1)  # Shape: (batch, 1, features)
            h = tf.matmul(x_exp, w1) + b1  # (batch, 1, 64)
            return tf.nn.relu(h)

        h = Lambda(compute_hidden)([x_in, w1, b1])

        # Compute output = h @ w2 + b2
        def compute_output(inputs):
            h, w2, b2 = inputs
            output = tf.matmul(h, w2) + b2  # (batch, 1, features)
            return tf.squeeze(output, axis=1)  # Remove extra dim

        output = Lambda(compute_output)([h, w2, b2])

        return Model(x_in, output)

    # Build and compile model
    hypernet = build_hypernetwork()
    final_model = build_final_model(hypernet)
    final_model.compile(optimizer='adam', loss='mse')

    # Meta-training
    for outer_epoch in range(outer_epochs):
        for i in range(len(X)):
            obs = mask[i] == 1
            if np.sum(obs) > 0:
                final_model.fit(X[i:i+1], X[i:i+1],
                                epochs=inner_epochs, verbose=0)

        final_model.fit(X, X, epochs=1, verbose=0)

    # Imputation
    X_hat = final_model.predict(X, verbose=0)
    X_imputed = mask * X + (1 - mask) * X_hat

    # Post-processing
    X_imputed = X_imputed * (max_val - min_val) + min_val
    X_imputed = np.round(X_imputed).astype(int)
    for i, cat in enumerate(encoder.categories_):
        X_imputed[:, i] = np.clip(X_imputed[:, i], 0, len(cat)-1)

    return pd.DataFrame(encoder.inverse_transform(X_imputed), columns=df_missing.columns)


def impute_neural_algorithm(df_missing, steps=10, epochs=50):
    import numpy as np
    import pandas as pd
    from sklearn.preprocessing import OrdinalEncoder
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Dense
    from tensorflow.keras.optimizers import Adam

    df = df_missing.copy()

    # Separate numerical and categorical columns
    num_cols = df.select_dtypes(include=[np.number]).columns
    cat_cols = df.select_dtypes(exclude=[np.number]).columns

    # Impute numeric columns with column mean (temporary)
    df[num_cols] = df[num_cols].astype(float)
    num_mask = df[num_cols].isna()
    df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

    # Encode categorical columns
    cat_filled = df[cat_cols].fillna('missing')
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    cat_encoded = encoder.fit_transform(cat_filled)

    # Combine numeric and encoded categorical columns
    X = np.hstack([df[num_cols].values, cat_encoded])
    X = X.astype(float)

    # Create mask: 1 = observed, 0 = missing
    full_mask = np.hstack([
        (~num_mask).astype(float).values,
        (cat_filled != 'missing').astype(float).values
    ])

    # Normalize
    min_val, max_val = np.min(X), np.max(X)
    X_norm = (X - min_val) / (max_val - min_val)

    # Replace nan with 0 for processing (should be none now)
    X_norm = np.nan_to_num(X_norm)

    # Neural processor model
    processor = Sequential([
        Dense(64, activation='relu'),
        Dense(64, activation='relu'),
        Dense(X.shape[1])
    ])
    processor.compile(optimizer=Adam(), loss='mse')

    # Iterative refinement
    X_current = X_norm.copy()
    for step in range(steps):
        delta = processor.predict(X_current, verbose=0)
        X_current = X_current + (1 - full_mask) * delta
        processor.fit(X_norm, X_current, epochs=epochs // steps, verbose=0)

    # Final imputed values
    X_imputed = full_mask * X_norm + (1 - full_mask) * X_current
    X_imputed = X_imputed * (max_val - min_val) + min_val

    # Decode categorical part
    imputed_df = pd.DataFrame(X_imputed, columns=list(num_cols) + list(cat_cols))
    for i, col in enumerate(cat_cols):
        cat_idx = len(num_cols) + i
        col_vals = np.clip(np.round(imputed_df.iloc[:, cat_idx]).astype(int), 0, len(encoder.categories_[i]) - 1)
        imputed_df[col] = encoder.inverse_transform(col_vals.values.reshape(-1, 1)).flatten()

    # Cast numerical columns back to original dtype
    for col in num_cols:
        imputed_df[col] = pd.to_numeric(imputed_df[col], errors='coerce')

    return imputed_df[df.columns]  # Ensure original column order
```

```python
def graph_attention_impute(df_missing, n_heads=4, epochs=100, lr=0.01):
    try:
        # 1. Create graph from correlations
        corr_matrix = df_missing.corr().abs()
        G = nx.from_pandas_adjacency(corr_matrix > 0.5)  # Threshold = 0.5

        # 2. Get adjacency matrix
        A = nx.adjacency_matrix(G).astype('f4')
        A_sp = sp_matrix_to_sp_tensor(A)  # Use sparse tensor!

        # 3. Prepare node features (normalized)
        X = df_missing.fillna(0).values
        X = (X - X.mean(0)) / (X.std(0) + 1e-8)

        # 4. Build GAT model
        N_nodes = X.shape[0]
        F_features = X.shape[1]

        X_in = Input(shape=(F_features,))
        A_in = Input((N_nodes,), sparse=True)

        gat = GATConv(64, attn_heads=n_heads, concat_heads=True)([X_in, A_in])
        out = Dense(F_features)(gat)

        model = Model(inputs=[X_in, A_in], outputs=out)
        model.compile(optimizer=Adam(lr), loss='mae')

        # 5. Create masks for observed values
        mask = ~np.isnan(df_missing.values)

        # 6. Train
        model.fit([X, A_sp],
                  X,
                  sample_weight=mask.astype(float),
                  epochs=epochs,
                  verbose=0)

        # 7. Impute missing values
        imputed = model.predict([X, A_sp], verbose=0)
        return pd.DataFrame(imputed,
                            columns=df_missing.columns,
                            index=df_missing.index)

    except Exception as e:
        print(f"GraphAttention failed: {str(e)}")
        print("Falling back to KNN imputation...")

        return pd.DataFrame(
            KNNImputer().fit_transform(df_missing),
            columns=df_missing.columns
        )


def impute_graph_attention(df_missing, epochs=50):
    """
    Graph Attention Network-based imputation.
    Treats features as nodes in a fully-connected graph.
    """
    import numpy as np
    import pandas as pd
    from sklearn.preprocessing import OrdinalEncoder
    import tensorflow as tf
    from tensorflow.keras import Model, Input
    from tensorflow.keras.layers import Dense
    from spektral.layers import GATConv
    from spektral.data import Graph

    # Encode categorical values
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Normalize to [0, 1]
    min_val = np.nanmin(encoded)
    max_val = np.nanmax(encoded)
    X = (encoded - min_val) / (max_val - min_val)

    # Create mask for missing
    mask = (~np.isnan(encoded)).astype(float)
    X = np.nan_to_num(X, nan=0)

    # Graph setup
    n_samples, n_features = X.shape
    # Treat features as nodes: node features are the observations
    X_nodes = X.T  # (n_features, n_samples)

    # Adjacency matrix: fully connected
    adj = np.ones((n_features, n_features)) - np.eye(n_features)  # Remove self-loops
    adj = adj.astype(np.float32)

    # Model inputs
    X_in = Input(shape=(n_samples,), name="X_in")  # each node has a feature vector of length n_samples
    A_in = Input(shape=(n_features,), name="A_in")

    # Graph Attention layers
    gat1 = GATConv(32, activation='relu')([X_in, A_in])
    gat2 = GATConv(32, activation='relu')([gat1, A_in])

    # Final reconstruction layer (reconstruct node features)
    output = Dense(n_samples, activation='sigmoid')(gat2)

    model = Model(inputs=[X_in, A_in], outputs=output)
    model.compile(optimizer='adam', loss='mse')

    # Fit the model
    model.fit([X_nodes, adj], X_nodes,
              epochs=epochs,
              batch_size=n_features,  # each node as a batch
              verbose=0)

    # Predict imputed features
    X_hat_nodes = model.predict([X_nodes, adj], verbose=0)
    X_hat = X_hat_nodes.T  # back to (n_samples, n_features)

    # Impute
    X_imputed = mask * X + (1 - mask) * X_hat

    # Denormalize
    X_imputed = X_imputed * (max_val - min_val) + min_val
    X_imputed = np.round(X_imputed).astype(int)
    X_imputed = np.clip(X_imputed, 0, len(encoder.categories_[0])-1)

    # Decode back to original categories
    imputed_df = pd.DataFrame(encoder.inverse_transform(X_imputed), columns=df_missing.columns)

    return imputed_df


#5. Spatial-Temporal Graph Networks
def impute_stgn(df_missing, epochs=50):
    """Spatial-Temporal Graph Network for imputation"""
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Normalize
    min_val = np.nanmin(encoded)
    max_val = np.nanmax(encoded)
    X = (encoded - min_val) / (max_val - min_val)
    X = np.nan_to_num(X, nan=0)
    mask = (~np.isnan(encoded)).astype(float)

    # Create temporal edges (sequential)
    adj = np.zeros((len(X), len(X)))
    for i in range(len(X)-1):
        adj[i,i+1] = 1
        adj[i+1,i] = 1

    # Build ST-GN model
    X_in = Input(shape=(X.shape[1],))
    A_in = Input(shape=(len(X),))

    # Temporal encoding
    temp = Dense(32)(A_in)
    temp = BatchNormalization()(temp)

    # Feature encoding
    feat = Dense(32)(X_in)
    feat = BatchNormalization()(feat)

    # Combine
    combined = Concatenate()([feat, temp])
    x = Dense(64, activation='relu')(combined)
    x = Dense(64, activation='relu')(x)
    outputs = Dense(X.shape[1], activation='sigmoid')(x)

    model = Model(inputs=[X_in, A_in], outputs=outputs)
    model.compile(optimizer='adam', loss='mse')

    # Train
    model.fit([X, adj], X, epochs=epochs, batch_size=32, verbose=0)

    # Impute
    X_hat = model.predict([X, adj], verbose=0)
    X_imputed = mask * X + (1 - mask) * X_hat

    # Convert back
    X_imputed = X_imputed * (max_val - min_val) + min_val
    X_imputed = np.round(X_imputed).astype(int)
    X_imputed = np.clip(X_imputed, 0, len(encoder.categories_[0])-1)

    return pd.DataFrame(encoder.inverse_transform(X_imputed), columns=df_missing.columns)
```

```python
from copulas.multivariate import GaussianMultivariate
from sklearn.preprocessing import OrdinalEncoder
import numpy as np
import pandas as pd


!pip install qiskit

!pip install "qiskit==0.39.5"


def impute_copula(df_missing):
    """Copula-based imputation using synthetic data"""

    df_copy = df_missing.copy()
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

    # Encode categorical values
    encoded = pd.DataFrame(encoder.fit_transform(df_copy), columns=df_copy.columns)

    # Use only complete rows to train the model
    complete_rows = encoded.dropna()
    model = GaussianMultivariate()
    model.fit(complete_rows)

    # Generate synthetic data of the same shape
    synthetic = pd.DataFrame(model.sample(len(encoded)), columns=encoded.columns)

    # Fill missing values
    imputed = encoded.copy()
    for col in imputed.columns:
        missing_idx = imputed[col].isna()
        imputed.loc[missing_idx, col] = synthetic.loc[missing_idx, col]

    # Decode back to original format
    imputed = np.round(imputed).astype(int)
    imputed = np.clip(imputed, 0, len(encoder.categories_[0]) - 1)
    df_imputed = pd.DataFrame(encoder.inverse_transform(imputed), columns=df_missing.columns)

    return df_imputed

from sklearn.cluster import KMeans

def impute_fuzzy_kmeans(df_missing, n_clusters=5):
    """Fuzzy K-Means imputation with soft assignments"""
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Get complete cases for initialization
    complete_mask = (~np.isnan(encoded)).all(axis=1)
    X_complete = encoded[complete_mask]

    if len(X_complete) == 0:
        return df_missing.fillna(df_missing.mode().iloc[0])

    # Initialize K-Means
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X_complete)

    # Get cluster centers and distances
    centers = kmeans.cluster_centers_
    distances = np.zeros((len(encoded), n_clusters))

    # Compute distances (handling missing values)
    for i in range(len(encoded)):
        for k in range(n_clusters):
            # Only use observed features for distance calculation
            observed = ~np.isnan(encoded[i])
            if np.sum(observed) > 0:
                distances[i,k] = np.sum((encoded[i,observed] - centers[k,observed])**2)
            else:
                distances[i,k] = np.inf

    # Compute membership probabilities
    epsilon = 1e-10
    weights = 1/(distances + epsilon)
    weights = weights / weights.sum(axis=1, keepdims=1)

    # Impute missing values
    imputed = encoded.copy()
    for i in range(len(encoded)):
        missing = np.isnan(encoded[i])
        if missing.any():
            # Weighted average of cluster centers
            imputed[i,missing] = np.sum(weights[i,:,np.newaxis] * centers[:,missing], axis=0)

    # Convert back
    imputed = np.round(imputed).astype(int)
    imputed = np.clip(imputed, 0, len(encoder.categories_[0])-1)

    return pd.DataFrame(encoder.inverse_transform(imputed), columns=df_missing.columns)

from sklearn.decomposition import NMF
from sklearn.impute import SimpleImputer

def matrix_factorization_impute(df_missing, n_components=5, max_iter=1000):
    """
    Matrix factorization imputation with dimension handling.
    """
    try:
        # 1. Initial imputation for missing values
        init_imputer = SimpleImputer(strategy='mean')
        X = init_imputer.fit_transform(df_missing)

        # 2. Validate component size
        valid_components = min(n_components, X.shape[0]-1, X.shape[1]-1)
        if valid_components < 1:
            valid_components = 1

        # 3. Perform Non-negative Matrix Factorization
        model = NMF(n_components=valid_components,
                   max_iter=max_iter,
                   init='random',
                   random_state=42)

        W = model.fit_transform(X)
        H = model.components_

        # 4. Reconstruct matrix
        reconstructed = np.dot(W, H)

        # 5. Fill missing values while preserving observed ones
        mask = df_missing.isna()
        imputed = df_missing.copy()
        imputed[mask] = reconstructed[mask]

        return imputed

    except Exception as e:
        print(f"Matrix Factorization failed: {str(e)}")
        print("Falling back to iterative imputer...")
        from sklearn.experimental import enable_iterative_imputer
        from sklearn.impute import IterativeImputer
        return pd.DataFrame(
            IterativeImputer().fit_transform(df_missing),
            columns=df_missing.columns
        )

def impute_matrix_factorization(df_missing, side_info=None, rank=5, epochs=50):

    import numpy as np
    import pandas as pd
    from sklearn.preprocessing import OrdinalEncoder

    # Encode categorical values
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Normalize to [0, 1]
    min_val = np.nanmin(encoded)
    max_val = np.nanmax(encoded)
    X = (encoded - min_val) / (max_val - min_val)
    mask = (~np.isnan(encoded)).astype(float)

    # Replace NaNs with 0s for calculation
    X = np.nan_to_num(X, nan=0)

    n_samples, n_features = X.shape

    # Prepare side info
    if side_info is None:
        side_info = np.zeros((n_samples, 0))  # No side info

    n_side = side_info.shape[1]

    # Initialize factor matrices
    U = np.random.normal(scale=0.1, size=(n_samples, rank))
    V = np.random.normal(scale=0.1, size=(n_features, rank))
    W = np.random.normal(scale=0.1, size=(n_side, rank)) if n_side > 0 else np.zeros((0, rank))

    # Training
    for epoch in range(epochs):
        # Update U
        for i in range(n_samples):
            observed = mask[i] == 1
            if np.sum(observed) > 0:
                V_obs = V[observed]
                X_obs = X[i, observed]
                side_term = W.T @ side_info[i] if n_side > 0 else 0
                U[i] = np.linalg.solve(
                    V_obs.T @ V_obs + np.eye(rank),
                    V_obs.T @ X_obs + side_term
                )

        # Update V
        for j in range(n_features):
            observed = mask[:, j] == 1
            if np.sum(observed) > 0:
                U_obs = U[observed]
                X_obs = X[observed, j]
                V[j] = np.linalg.solve(
                    U_obs.T @ U_obs + np.eye(rank),
                    U_obs.T @ X_obs
                )

        # Update W if side info is present
        if n_side > 0:
            W = np.linalg.solve(
                U.T @ U + np.eye(rank),
                U.T @ side_info
            )

    # Impute missing values
    X_hat = U @ V.T
    X_imputed = mask * X + (1 - mask) * X_hat

    # Convert back to original scale
    X_imputed = X_imputed * (max_val - min_val) + min_val
    X_imputed = np.round(X_imputed).astype(int)
    X_imputed = np.clip(X_imputed, 0, len(encoder.categories_[0]) - 1)

    # Decode back to original categories
    imputed_df = pd.DataFrame(encoder.inverse_transform(X_imputed), columns=df_missing.columns)

    return imputed_df




def ntm_impute(df_missing, epochs=100, batch_size=32):
    """
    Neural Tensor Modeling (NTM) imputation using LSTM.
    Imputes missing values only; keeps original non-missing values.
    """
    try:
        # 1. Preprocess data
        scaler = MinMaxScaler()
        scaled = scaler.fit_transform(df_missing.fillna(0))
        mask = (~df_missing.isna()).astype(int).values

        # 2. 3D input shape
        X = np.expand_dims(scaled, axis=1)
        mask = np.expand_dims(mask, axis=1)

        # 3. LSTM model
        model = Sequential([
            Masking(mask_value=0, input_shape=(1, X.shape[2])),
            LSTM(64, return_sequences=True),
            Dropout(0.2),
            Dense(X.shape[2])
        ])
        model.compile(optimizer='adam', loss='mse')

        # 4. Train
        model.fit(
            X * mask,
            X,
            epochs=epochs,
            batch_size=batch_size,
            callbacks=[EarlyStopping(patience=5)],
            verbose=0
        )

        # 5. Impute
        imputed = model.predict(X)
        imputed = scaler.inverse_transform(imputed[:,0,:])

        # 6. Restore observed values
        imputed_df = pd.DataFrame(imputed, columns=df_missing.columns)
        imputed_df[df_missing.notna()] = df_missing[df_missing.notna()]

        return imputed_df

    except Exception as e:
        print(f"NTM imputation failed: {str(e)}")
        return df_missing


from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, RepeatVector, Masking, Dense
from tensorflow.keras.callbacks import EarlyStopping


def impute_lstm_autoencoder(df_missing, epochs=100):
    df_copy = df_missing.copy()
    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(df_copy.fillna(0))
    mask = (~df_copy.isna()).astype(float).values

    X_scaled = np.expand_dims(X_scaled, axis=1)

    # Build simple LSTM autoencoder
    inputs = Input(shape=(1, X_scaled.shape[2]))
    encoded = LSTM(64, return_sequences=True)(inputs)
    decoded = LSTM(X_scaled.shape[2], return_sequences=True)(encoded)
    autoencoder = Model(inputs, decoded)
    autoencoder.compile(optimizer='adam', loss='mse')

    # Train only on observed values
    autoencoder.fit(
        X_scaled * mask[:, np.newaxis, :],
        X_scaled * mask[:, np.newaxis, :],
        epochs=epochs,
        verbose=0
    )

    # Predict
    X_pred = autoencoder.predict(X_scaled)

    # Replace only missing values
    X_imputed = mask[:, np.newaxis, :] * X_scaled + (1 - mask[:, np.newaxis, :]) * X_pred

    # Rescale
    X_imputed_final = scaler.inverse_transform(X_imputed[:, 0, :])
    df_imputed = pd.DataFrame(X_imputed_final, columns=df_missing.columns)

    return df_imputed




def impute_ntm(df_missing, epochs=50):

  try:
    """Neural Turing Machine for sequential imputation"""
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Normalize
    min_val = np.nanmin(encoded)
    max_val = np.nanmax(encoded)
    X = (encoded - min_val) / (max_val - min_val)
    X = np.nan_to_num(X, nan=0)
    mask = (~np.isnan(encoded)).astype(float)

    # Reshape for sequence processing
    X_seq = X.reshape(1, *X.shape)  # (1, seq_len, features)
    mask_seq = mask.reshape(1, *mask.shape)

    # Build NTM-like model
    inputs = Input(shape=(X.shape[0], X.shape[1]))
    mask_input = Input(shape=(X.shape[0], X.shape[1]))

    # Memory component
    memory = tf.zeros((1, 128, 20))  # (batch, memory_size, word_size)

    # Controller (LSTM)
    lstm = LSTM(64, return_sequences=True)(inputs)

    # Read/write mechanisms (simplified)
    read = Dense(20)(lstm)
    write = Dense(20)(lstm)

    # Update memory
    memory = memory + tf.expand_dims(write, axis=1)

    # Output
    output = Dense(X.shape[1], activation='sigmoid')(
        Concatenate()([lstm, tf.reduce_mean(memory, axis=1)]))

    model = Model(inputs=[inputs, mask_input], outputs=output)
    model.compile(optimizer='adam', loss='mse')

    # Train
    model.fit([X_seq, mask_seq], X_seq, epochs=epochs, verbose=0)

    # Impute
    X_hat = model.predict([X_seq, mask_seq], verbose=0)[0]
    X_imputed = mask * X + (1 - mask) * X_hat

    # Convert back
    X_imputed = X_imputed * (max_val - min_val) + min_val
    X_imputed = np.round(X_imputed).astype(int)
    X_imputed = np.clip(X_imputed, 0, len(encoder.categories_[0])-1)

    return pd.DataFrame(encoder.inverse_transform(X_imputed), columns=df_missing.columns)

  except Exception as e:
        print(f"Neural Turing Machine for sequential imputation failed: {str(e)}")
        return df_missing  # Return original if fails

def quantum_impute(df_missing, n_qubits=4, max_iter=100):
    """
    Quantum circuit-based imputation using parameterized circuits.

    """
    try:
        # 1. Normalize data
        scaler = MinMaxScaler()
        scaled = scaler.fit_transform(df_missing.fillna(0))
        mask = (~df_missing.isna()).astype(int)

        # 2. Create parameterized quantum circuit
        params = [Parameter(f'θ_{i}') for i in range(n_qubits)]
        qc = QuantumCircuit(n_qubits)

        for i in range(n_qubits):
            qc.ry(params[i], i)

        qc.barrier()
        for i in range(n_qubits-1):
            qc.cx(i, i+1)

        # 3. Define quantum cost function
        def cost_function(params_values):
            # Bind parameters
            bound_qc = qc.bind_parameters(dict(zip(params, params_values)))

            # Simulate
            simulator = Aer.get_backend('statevector_simulator')
            result = execute(bound_qc, simulator).result()
            statevector = np.abs(result.get_statevector())**2

            # Calculate loss (compare with observed values)
            loss = np.sum(mask * (statevector[:len(scaled)] - scaled)**2)
            return loss

        # 4. Optimize parameters
        optimizer = COBYLA(maxiter=max_iter)
        initial_params = np.random.uniform(0, 2*np.pi, len(params))
        opt_result = optimizer.minimize(cost_function, initial_params)

        # 5. Generate imputations
        bound_qc = qc.bind_parameters(dict(zip(params, opt_result.x)))
        simulator = Aer.get_backend('statevector_simulator')
        result = execute(bound_qc, simulator).result()
        imputed = np.abs(result.get_statevector()[:len(scaled)])**2

        # 6. Inverse transform
        imputed = scaler.inverse_transform(imputed.reshape(-1, 1))
        return pd.DataFrame(imputed, columns=df_missing.columns)

    except Exception as e:
        print(f"Quantum imputation failed: {str(e)}")
        print("Falling back to mean imputation...")
        return df_missing.fillna(df_missing.mean())


# Placeholder for the Qiskit Iotimes operation needed in objective_function (requires qiskit.opflow)
# This is a basic utility function; a proper Qiskit setup might handle this differently.
from qiskit.opflow import I, PauliOp
def Iotimes(n):
    """Returns the identity operator tensored n times."""
    if n == 0:
        return I
    op = I
    for _ in range(n - 1):
        op = op ^ I
    return op

```

```python
# --- Impute Quantum Function ---
def impute_quantum(df_missing, n_qubits=None, max_iter_vqe=100, vqe_optimizer=None):

    print("Attempting quantum imputation...")

    try:
        # --- Data Preparation ---
        # Ensure data is numeric. Ordinal encode categorical features.
        df_temp = df_missing.copy()
        cat_cols = df_temp.select_dtypes(include=['object', 'category']).columns
        num_cols = df_temp.select_dtypes(include=np.number).columns

        temp_encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

        if len(cat_cols) > 0:
            # Encode categorical columns
            df_temp[cat_cols] = temp_encoder.fit_transform(df_temp[cat_cols])

        # Convert all data to float for the quantum process
        df_temp_numeric = df_temp.apply(pd.to_numeric, errors='coerce')

        # Handle columns that became all NaN after coercion (e.g., if they were originally purely missing categorical)
        for col in df_temp_numeric.columns:
            if np.all(np.isnan(df_temp_numeric[col])):
                 print(f"Warning: Column '{col}' became all NaN during numeric conversion for quantum. Filling with 0.")
                 df_temp_numeric[col] = 0 # Or some other default

        # Store column order and types for inverse transformation
        original_columns = df_missing.columns
        original_dtypes = df_missing.dtypes
        column_is_categorical = {col: col in cat_cols for col in original_columns}

        # Normalize data for quantum circuit input (often [-1, 1] or [0, 1])
        # Let's use [0, 1] for simplicity with a sigmoid output ansatz
        min_vals = df_temp_numeric.min(axis=0)
        max_vals = df_temp_numeric.max(axis=0)
        ranges = max_vals - min_vals
        # Avoid division by zero for constant columns
        ranges[ranges == 0] = 1e-8 # Use a small epsilon

        data_normalized = (df_temp_numeric - min_vals) / ranges
        # Fill NaNs AFTER normalization for data fed to the circuit/objective
        # Missing values will be the ones we try to impute. We need a mask.
        mask_np = (~df_missing.isna()).astype(float) # 1 if observed, 0 if missing

        # The data fed into the objective should distinguish observed from missing.
        # A common approach is to feed the normalized data with missing filled (e.g., 0)
        # and the mask.
        data_normalized_filled = np.nan_to_num(data_normalized.values, nan=0.0)
        mask_values = mask_np.values

        n_features = data_normalized_filled.shape[1]

        if n_qubits is None:
             # Simple heuristic: use min(number of features, a practical limit)
             n_qubits = min(n_features, 6) # Cap at 6 for simulator performance on complex circuits

        if n_qubits > n_features:
             print(f"Warning: n_qubits ({n_qubits}) > number of features ({n_features}). Using n_qubits = n_features.")
             n_qubits = n_features
        if n_qubits <= 0:
             raise ValueError("n_qubits must be a positive integer.")


        print(f"Using {n_qubits} qubits for quantum imputation.")

        if n_qubits > 5:
            print(f"Warning: Using {n_qubits} qubits might be computationally heavy. Ensure enough resources.")


        qc = QuantumCircuit(n_qubits)
        # We need parameters for each rotation. A simple ansatz might use 2 parameters per qubit.
        num_params = n_qubits * 2 # Example: Ry(theta1) -> CZ -> Ry(theta2)
        params = ParameterVector('θ', num_params)

        # Initial layer of rotations
        for i in range(n_qubits):
            qc.ry(params[i], i)
        qc.barrier()

        # Entangling layer (linear or circular)
        for i in range(n_qubits - 1):
             qc.cz(i, i + 1)
        # Add connection between last and first qubit for circular entanglement
        if n_qubits > 1:
             qc.cz(n_qubits - 1, 0)
        qc.barrier()

        # Second layer of rotations
        for i in range(n_qubits):
             qc.ry(params[n_qubits + i], i)

        observables = []
        for i in range(n_qubits):
            # Create a Pauli string like 'II...Z...II' where Z is on qubit i
            pauli_string = ['I'] * n_qubits
            pauli_string[i] = 'Z'
            observables.append(SparsePauliOp.from_list(["".join(pauli_string)]))


        estimator = AerEstimator() # Use AerSimulator backend for estimation


        def vqe_objective(param_values):

            # Bind parameters to the quantum circuit
            bound_qc = qc.assign_parameters(dict(zip(params, param_values)))

            # Calculate expectation values for each observable (each qubit)
            # The result of run() is a Job, fetch results later
            job = estimator.run([bound_qc] * len(observables), observables)
            try:
                 exp_values = job.result().values # Get expectation values
            except Exception as e:
                 print(f"Error during Qiskit estimation: {e}")
                 # Return a high loss to penalize failed evaluations
                 return 1e10

            predicted_normalized_values = (np.array(exp_values) + 1) / 2.0

            batch_indices = np.random.choice(data_normalized_filled.shape[0], size=min(batch_size, data_normalized_filled.shape[0]), replace=False)
            batch_data = data_normalized_filled[batch_indices, :n_qubits] # Only compare first n_qubits features
            batch_mask = mask_values[batch_indices, :n_qubits] # Mask for these features


            if n_features < n_qubits:
                 raise ValueError("Cannot match expectation values to features if n_qubits > n_features")

            # Calculate the mean of the observed data for the first n_qubits features
            # Need to handle NaNs correctly when calculating the mean
            observed_means = np.nanmean(df_temp_numeric.values[:, :n_qubits], axis=0)
            # Normalize the observed means the same way the data was normalized
            observed_means_normalized = (observed_means - min_vals.values[:n_qubits]) / ranges.values[:n_qubits]

            # If a feature was all NaN, its normalized mean will be NaN.
            # Replace NaN means with 0 or the overall mean? Let's use 0 for now.
            observed_means_normalized = np.nan_to_num(observed_means_normalized, nan=0.0)


            # Compare the predicted expectation values (scaled to [0,1]) to the observed means
            # Use only the first n_qubits predictions if circuit outputs more
            predicted_for_comparison = predicted_normalized_values[:n_qubits]

            # Calculate Mean Squared Error between predicted and observed means
            loss = np.mean((predicted_for_comparison - observed_means_normalized) ** 2)


            return loss

        # --- Run the VQE Optimization ---
        initial_params = np.random.rand(num_params) * 2 * np.pi # Random initial parameters
        if vqe_optimizer is None:
             vqe_optimizer = COBYLA(maxiter=max_iter_vqe)

        print(f"Starting VQE optimization with {type(vqe_optimizer).__name__} for {max_iter_vqe} iterations...")
        # Minimize the objective function
        result = vqe_optimizer.minimize(vqe_objective, initial_params)
        optimal_params = result.x
        print(f"VQE optimization finished. Success: {result.success}, Message: {result.message}")

        bound_qc_optimal = qc.assign_parameters(dict(zip(params, optimal_params)))
        job_optimal = estimator.run([bound_qc_optimal] * len(observables), observables)
        try:
             learned_normalized_template = (np.array(job_optimal.result().values) + 1) / 2.0
             # Only take values corresponding to features if n_qubits <= n_features
             learned_normalized_template = learned_normalized_template[:n_features]
        except Exception as e:
             print(f"Error getting expectation values from optimal parameters: {e}")
             raise RuntimeError("Failed to get learned template from quantum circuit.") from e


        # Scale the template back to the original data range
        learned_template = learned_normalized_template * ranges.values + min_vals.values

        imputed_df_numeric = df_temp_numeric.copy()

        for col_index in range(n_features):
             col_name = original_columns[col_index] # Get original column name
             if np.isnan(df_missing[col_name]).any(): # Check if original column had any NaNs
                  # Find indices where the original column was missing
                  missing_indices = df_missing[col_name].index[df_missing[col_name].isna()]
                  # Fill these missing entries with the learned template value for this feature
                  imputed_df_numeric.loc[missing_indices, col_name] = learned_template[col_index]


        # --- Inverse Transform Categorical Columns ---
        imputed_df_final = imputed_df_numeric.copy() # Start with the numeric imputed data

        if len(cat_cols) > 0:
             for i, col_name in enumerate(cat_cols):
                 # Ensure the column exists in the imputed numeric data
                 if col_name in imputed_df_final.columns:
                    # Get the imputed numeric values for this original categorical column
                    col_imputed_values = imputed_df_final[col_name].values

                    # Find the index of this column in the encoder's categories list (used during fit)
                    try:
                        encoder_col_index = list(df_missing.columns).index(col_name) # Index in original columns
                        num_categories = len(temp_encoder.categories_[encoder_col_index])

                        # Round and clip the imputed numeric values to valid encoded integer indices
                        rounded_values = np.round(col_imputed_values)
                        imputed_encoded_clipped_col = np.clip(rounded_values, 0, num_categories - 1).astype(int)

                        # Inverse transform this column using the encoder fitted earlier
                        # Inverse transform expects a 2D array shape (n_samples, n_features_this_transform)
                        imputed_categorical_original_col = temp_encoder.inverse_transform(imputed_encoded_clipped_col.reshape(-1, 1))

                        # Replace the numeric column in the final DataFrame with the original categorical values
                        imputed_df_final[col_name] = imputed_categorical_original_col.flatten() # Use flatten to get 1D array


                    except IndexError:
                        print(f"Warning: Category index mismatch for column {col_name} during quantum inverse transform. Skipping inverse transform for this column.")
                        # Fallback: keep as numeric or fill with mode if conversion failed
                        mode_val = df_missing[col_name].mode()[0] if not df_missing[col_name].mode().empty else np.nan
                        imputed_df_final[col_name] = imputed_df_final[col_name].fillna(mode_val)
                    except Exception as inv_e:
                        print(f"Error during inverse transform for column {col_name} after quantum imputation: {inv_e}")
                        # Fallback: keep as numeric or fill with mode
                        mode_val = df_missing[col_name].mode()[0] if not df_missing[col_name].mode().empty else np.nan
                        imputed_df_final[col_name] = imputed_df_final[col_name].fillna(mode_val)

        # Ensure non-categorical columns are numerical
        for col in num_cols:
             if col in imputed_df_final.columns:
                 imputed_df_final[col] = pd.to_numeric(imputed_df_final[col], errors='coerce')
                 # Fill any NaNs that might have been introduced by coercion or if the template was NaN
                 imputed_df_final[col] = imputed_df_final[col].fillna(df_missing[col].median() if df_missing[col].median() else 0)


        # Reindex to ensure exact column order and presence as original
        imputed_df_final = imputed_df_final.reindex(columns=original_columns)

        print("Quantum imputation completed.")
        return imputed_df_final

    except Exception as e:
        print(f"Quantum imputation failed: {str(e)}")
        print("Falling back to IterativeImputer...")
        try:
            # Fallback mechanism (Robust version)
            df_temp = df_missing.copy()

            # Check if there are any missing values at all before encoding
            if df_temp.isna().sum().sum() == 0:
                print("No missing values found. Returning original DataFrame.")
                return df_temp

            # Ordinal encode categorical features first
            cat_cols = df_temp.select_dtypes(include=['object', 'category']).columns
            num_cols = df_temp.select_dtypes(include=np.number).columns

            temp_encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

            # Apply encoder only to categorical columns
            df_temp_encoded = df_temp.copy()
            if len(cat_cols) > 0:
                # Fit and transform only categorical columns
                df_temp_encoded[cat_cols] = temp_encoder.fit_transform(df_temp[cat_cols])

            # Convert all columns to numeric type before Imputer, coercing errors
            df_temp_numeric = df_temp_encoded.apply(pd.to_numeric, errors='coerce')

            # Handle columns that might have become all NaN after coercion
            for col in df_temp_numeric.columns:
                if np.all(np.isnan(df_temp_numeric[col])):
                     print(f"Warning: Column '{col}' became all NaN during numeric conversion for fallback. Filling with 0.")
                     df_temp_numeric[col] = 0

            imputer_fallback = IterativeImputer(random_state=42)
            imputed_data_fallback_numeric = imputer_fallback.fit_transform(df_temp_numeric)

            # Convert back to DataFrame to preserve column names and index
            imputed_df_fallback_numeric = pd.DataFrame(imputed_data_fallback_numeric, columns=df_temp_numeric.columns, index=df_temp_numeric.index)

            # --- Apply Inverse Transform to Categorical Columns Only in Fallback ---
            imputed_df_fallback = imputed_df_fallback_numeric.copy()

            if len(cat_cols) > 0:
                 for i, col_name in enumerate(cat_cols):
                     if col_name in imputed_df_fallback.columns:
                         col_imputed_values = imputed_df_fallback[col_name].values
                         try:
                            # Find the index of this column in the encoder's categories list (used during fit)
                            encoder_col_index = list(df_missing.columns).index(col_name) # Index in original columns
                            num_categories = len(temp_encoder.categories_[encoder_col_index])

                            # Round and clip the imputed numeric values to valid encoded integer indices
                            rounded_values = np.round(col_imputed_values)
                            imputed_encoded_clipped_col = np.clip(rounded_values, 0, num_categories - 1).astype(int)

                            # Inverse transform this column using the encoder fitted earlier
                            imputed_categorical_original_col = temp_encoder.inverse_transform(imputed_encoded_clipped_col.reshape(-1, 1))
                            imputed_df_fallback[col_name] = imputed_categorical_original_col.flatten()

                         except IndexError:
                            print(f"Warning: Category index mismatch for column {col_name} during fallback inverse transform. Skipping inverse transform for this column.")
                            mode_val = df_missing[col_name].mode()[0] if not df_missing[col_name].mode().empty else np.nan
                            imputed_df_fallback[col_name] = imputed_df_fallback[col_name].fillna(mode_val)
                         except Exception as inv_e:
                             print(f"Error during inverse transform for column {col_name} after fallback imputation: {inv_e}")
                             mode_val = df_missing[col_name].mode()[0] if not df_missing[col_name].mode().empty else np.nan
                             imputed_df_fallback[col_name] = imputed_df_fallback[col_name].fillna(mode_val)


            # Ensure non-categorical columns are numerical
            for col in num_cols:
                 if col in imputed_df_fallback.columns:
                     imputed_df_fallback[col] = pd.to_numeric(imputed_df_fallback[col], errors='coerce')
                     imputed_df_fallback[col] = imputed_df_fallback[col].fillna(df_missing[col].median() if df_missing[col].median() else 0)


            # Reindex to ensure exact column order and presence as original
            imputed_df_fallback = imputed_df_fallback.reindex(columns=original_columns)


            print("Fallback IterativeImputer completed successfully.")
            return imputed_df_fallback

        except Exception as fallback_e:
             print(f"Fallback IterativeImputer also failed: {str(fallback_e)}")
             # Final fallback to mode imputation
             print("Falling back to mode imputation...")
             return impute_mode(df_missing.copy())
```

```python
def impute_saits(df_missing, d_model=64, num_heads=4, epochs=50):
    """
    SAITS-like imputation: Self-Attention Imputation for Time Series.
    Uses self-attention layers to impute missing values.
    """
    import numpy as np
    import pandas as pd
    from sklearn.preprocessing import OrdinalEncoder
    import tensorflow as tf
    from tensorflow.keras import Model, Input
    from tensorflow.keras.layers import Dense, Embedding, LayerNormalization, Add
    from tensorflow.keras.layers import MultiHeadAttention

    # Encode categorical data
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Normalize
    min_val = np.nanmin(encoded)
    max_val = np.nanmax(encoded)
    X = (encoded - min_val) / (max_val - min_val)
    mask = (~np.isnan(encoded)).astype(float)
    X = np.nan_to_num(X, nan=0)

    n_samples, n_features = X.shape

    # Build SAITS-like model
    inputs = Input(shape=(n_features,))
    mask_input = Input(shape=(n_features,))

    # Add feature embedding (dense projection)
    x = Dense(d_model)(inputs)

    # Add positional encoding
    positions = tf.range(start=0, limit=n_features, delta=1)
    pos_enc = Embedding(input_dim=n_features, output_dim=d_model)(positions)
    pos_enc = tf.expand_dims(pos_enc, axis=0)  # shape (1, n_features, d_model)

    # Expand dims of input to (batch_size, n_features, d_model)
    x = tf.expand_dims(x, axis=1)  # shape (batch_size, 1, d_model)
    x = tf.repeat(x, repeats=n_features, axis=1)  # shape (batch_size, n_features, d_model)
    x = Add()([x, pos_enc])

    # Self-attention layers
    for _ in range(2):
        attn_output = MultiHeadAttention(num_heads=num_heads, key_dim=d_model)(x, x)
        x = LayerNormalization()(x + attn_output)
        ffn_output = Dense(d_model, activation='relu')(x)
        x = LayerNormalization()(x + ffn_output)

    # Output projection to original feature size
    outputs = Dense(1, activation='sigmoid')(x)
    outputs = tf.squeeze(outputs, axis=-1)  # (batch_size, n_features)

    model = Model(inputs=[inputs, mask_input], outputs=outputs)
    model.compile(optimizer='adam', loss='mse')

    # Train
    model.fit([X, mask], X,
              epochs=epochs,
              batch_size=32,
              verbose=0)

    # Impute
    X_hat = model.predict([X, mask], verbose=0)
    X_imputed = mask * X + (1 - mask) * X_hat

    # Denormalize and convert back
    X_imputed = X_imputed * (max_val - min_val) + min_val
    X_imputed = np.round(X_imputed).astype(int)
    X_imputed = np.clip(X_imputed, 0, len(encoder.categories_[0]) - 1)

    return pd.DataFrame(encoder.inverse_transform(X_imputed), columns=df_missing.columns)
```

```python
# ======================
# Bayesian PCA Imputation
# ======================
def impute_bayesian_pca(df_missing, n_components=2, max_iter=100, tol=1e-6):
    """Bayesian PCA imputation with robust error handling"""


    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

##2    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
    encoded = encoder.fit_transform(df_missing)

    # Initialize with mean imputation
    col_means = np.nanmean(encoded, axis=0)
    imputed = np.where(np.isnan(encoded), col_means, encoded)

    n_samples, n_features = encoded.shape
    n_components = min(n_components, n_features)

    # Initialize parameters
    W = np.random.randn(n_features, n_components) * 0.01
    sigma2 = 1.0
    alpha = np.ones(n_components)

    for iteration in range(max_iter):
        try:
            # E-step
            M = W.T @ W + sigma2 * np.eye(n_components)
            M_inv = np.linalg.inv(M)
            X = imputed @ W @ M_inv

            # M-step
            W_new = (imputed.T @ X) @ np.linalg.inv(X.T @ X + sigma2 * np.diag(alpha))
            sigma2_new = np.mean((imputed - X @ W.T) ** 2)

            # Check convergence
            if np.linalg.norm(W_new - W) < tol:
                break

            W = W_new
            sigma2 = sigma2_new

        except np.linalg.LinAlgError:
            sigma2 = max(sigma2 * 0.9, 1e-6)  # Reduce sigma2 if matrix is singular
            continue

    # Final imputation
    X_full = imputed @ W @ np.linalg.pinv(W.T @ W + sigma2 * np.eye(n_components))
    imputed_values = X_full @ W.T

    # Only replace missing values
    final_imputed = np.where(np.isnan(encoded), imputed_values, encoded)

    # Ensure valid categorical values
    final_imputed = np.round(final_imputed).astype(int)
    final_imputed = np.clip(final_imputed, 0, len(encoder.categories_[0])-1)

    return pd.DataFrame(encoder.inverse_transform(final_imputed),
                       columns=df_missing.columns)

# ======================
# Ensemble Imputation
# ======================
def impute_ensemble(df_missing, methods=None, voting='soft'):
    """Robust ensemble imputation combining multiple methods"""
    if methods is None:
        methods = [
            ('mode', impute_mode),
            ('knn', lambda x: impute_knn(x, n_neighbors=5)),
            ('iterative', impute_iterative),
            ('mice', impute_mice_forest)
        ]

    # Get imputations from all methods
    imputations = {}
    for name, method in methods:
        try:
            imputed = method(df_missing.copy())
            if imputed.isna().any().any():
                imputed = imputed.fillna(imputed.mode().iloc[0])
            imputations[name] = imputed
        except Exception as e:
            print(f"Skipping {name} due to error: {str(e)}")
            continue

    if not imputations:
        print("All ensemble methods failed, falling back to mode imputation")
        return impute_mode(df_missing)

    # Encode all results consistently
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    all_values = pd.concat(imputations.values()).values.reshape(-1, 1)
    encoder.fit(all_values)

    encoded_imps = {}
    for name, df in imputations.items():
        try:
            encoded = encoder.transform(df.values)
            encoded_imps[name] = encoded
        except Exception as e:
            print(f"Encoding failed for {name}: {str(e)}")
            continue

    if not encoded_imps:
        return impute_mode(df_missing)

    # Stack all imputations
    valid_methods = list(encoded_imps.keys())
    stacked = np.stack([encoded_imps[name] for name in valid_methods])

    # Handle voting scheme
    try:
        if voting == 'hard':
            imputed_encoded = np.round(np.nanmean(stacked, axis=0)).astype(int)
        else:
            # Soft voting with weights
            weights = {'mode': 0.2, 'knn': 0.3, 'iterative': 0.3, 'mice': 0.2}
            total_weight = sum(weights.get(m, 1.0/len(valid_methods)) for m in valid_methods)
            imputed_encoded = np.zeros_like(stacked[0])

            for i, name in enumerate(valid_methods):
                weight = weights.get(name, 1.0/len(valid_methods))
                imputed_encoded += np.nan_to_num(stacked[i], nan=0) * weight

            imputed_encoded = np.round(imputed_encoded / total_weight).astype(int)

        # Clip to valid category indices
        imputed_encoded = np.clip(imputed_encoded, 0, len(encoder.categories_[0])-1)

        # Convert back to original categories
        result_df = pd.DataFrame(encoder.inverse_transform(imputed_encoded),
                               columns=df_missing.columns)

        return result_df.fillna(result_df.mode().iloc[0])

    except Exception as e:
        print(f"Ensemble voting failed: {str(e)}, falling back to best single method")
        best_method = max(weights.items(), key=lambda x: x[1])[0]
        return imputations.get(best_method, impute_mode(df_missing))

# ======================
# Kalman Filter Imputation
# ======================
def impute_kalman(df_missing):
    """Kalman filter implementation for missing data imputation"""


    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

##7    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
    encoded = encoder.fit_transform(df_missing)

    imputed = encoded.copy()
    n_samples, n_features = encoded.shape

    for col in range(n_features):
        ts = encoded[:, col]
        mask = np.isnan(ts)

        if np.any(mask):
            try:
                # Initialize Kalman filter
                kf = KalmanFilter(
                    k_endog=1,  # Observation dimension
                    k_states=1,  # State dimension
                    initialization='diffuse'
                )

                # Simple random walk model
                kf['design'] = np.eye(1)
                kf['transition'] = np.eye(1)
                kf['selection'] = np.eye(1)
                kf['obs_cov'] = np.eye(1) * 0.1
                kf['state_cov'] = np.eye(1) * 0.1

                # Filter known values
                filtered_state = kf.filter(ts[~mask].reshape(-1, 1))

                # Smooth (impute missing values)
                smoothed_state = kf.smooth(ts[~mask].reshape(-1, 1))

                # Fill missing values with smoothed estimates
                imputed[mask, col] = smoothed_state.smoothed_state[0, mask]

            except Exception as e:
                print(f"Kalman failed for column {col}: {str(e)}")
                # Fallback to linear interpolation
                imputed[:, col] = np.interp(
                    np.arange(n_samples),
                    np.arange(n_samples)[~mask],
                    ts[~mask]
                )

    # Ensure proper categorical values
    imputed = np.round(imputed).astype(int)
    imputed = np.clip(imputed, 0, len(encoder.categories_[col])-1)

    return pd.DataFrame(encoder.inverse_transform(imputed),
                       columns=df_missing.columns)

# Add ARIMA imputation function
def impute_arima(df_missing, order=(1,0,1)):
    """ARIMA-based imputation for time-series-like categorical data"""

    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

##10    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
    encoded = encoder.fit_transform(df_missing)

    imputed = encoded.copy()
    n_samples, n_features = encoded.shape

    for col in range(n_features):
        ts = encoded[:, col]
        mask = np.isnan(ts)

        if np.any(mask):
            try:
                # Convert to pandas Series for ARIMA
                ts_series = pd.Series(ts[~mask])

                # Fit ARIMA model
                model = ARIMA(ts_series, order=order)
                model_fit = model.fit()

                # Get predictions for all indices
                predictions = model_fit.predict(start=0, end=len(ts)-1, typ='levels')

                # Fill missing values
                imputed[mask, col] = predictions[mask]

            except Exception as e:
                print(f"ARIMA failed for column {col}: {str(e)} - using linear interpolation")
                # Fallback to linear interpolation
                imputed[:, col] = np.interp(
                    np.arange(n_samples),
                    np.arange(n_samples)[~mask],
                    ts[~mask]
                )

    # Ensure proper categorical values
    imputed = np.round(imputed).astype(int)
    imputed = np.clip(imputed, 0, len(encoder.categories_[col])-1)

    return pd.DataFrame(encoder.inverse_transform(imputed),
                       columns=df_missing.columns)

```

```python
def impute_optimal_transport(df_missing, reg=0.1):
    """
    Optimal transport based imputation.
    Matches distributions of complete and incomplete cases.
    Optimal transport preserves distributions

    """
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Separate complete and incomplete cases
    complete_mask = (~np.isnan(encoded)).all(axis=1)
    X_complete = encoded[complete_mask]
    X_missing = encoded[~complete_mask]

    if len(X_complete) == 0:
        return df_missing.fillna(df_missing.mode().iloc[0])

    # Compute OT plan
    M = np.zeros((len(X_missing), len(X_complete)))
    for i in range(len(X_missing)):
        for j in range(len(X_complete)):
            # Simple distance metric - could be improved
            observed_mask = ~np.isnan(X_missing[i])
            M[i,j] = np.sum((X_missing[i,observed_mask] - X_complete[j,observed_mask])**2)



    # Compute transport plan
    a = np.ones(len(X_missing)) / len(X_missing)
    b = np.ones(len(X_complete)) / len(X_complete)
    a /= np.sum(a)  # Ensure sums match
    b /= np.sum(b)

    P = emd(a, b, M, numItermax=1000000)


    # Impute
    X_imputed = X_missing.copy()
    for i in range(len(X_missing)):
        # Find most similar complete cases
        top_k = np.argsort(P[i])[-3:]  # Take top 3
        for col in range(X_missing.shape[1]):
            if np.isnan(X_missing[i,col]):
                # Weighted average of top matches
                X_imputed[i,col] = np.sum(P[i,top_k] * X_complete[top_k,col]) / np.sum(P[i,top_k])

    # Combine back
    result = encoded.copy()
    result[~complete_mask] = X_imputed

    # Convert back
    result = np.round(result).astype(int)
    result = np.clip(result, 0, len(encoder.categories_[0])-1)

    return pd.DataFrame(encoder.inverse_transform(result), columns=df_missing.columns)
```

**4. Evaluation Metrics**

```python
def evaluate_imputation_accuracy(original, imputed, mask):
    """Calculate accuracy of imputed values compared to original"""
    correct = (original[mask] == imputed[mask]).sum().sum()
    total = mask.sum().sum()
    return correct / total if total else 0

# In evaluate_imputation_mae function:
def evaluate_imputation_mae(original, imputed, mask):
    """Calculate Mean Absolute Error only on imputed cells."""
    # Ensure inputs are pandas DataFrames
    original_df = pd.DataFrame(original)
    imputed_df = pd.DataFrame(imputed)

    # Ensure mask is a boolean DataFrame with the same shape
    mask_df = pd.DataFrame(mask)

    # Apply mask to select only imputed cells
    original_imputed_cells = original_df[mask_df]
    imputed_imputed_cells = imputed_df[mask_df]

    # Flatten the selected cells, coercing to numeric
    original_flat = pd.to_numeric(original_imputed_cells.values.flatten(), errors='coerce')
    imputed_flat = pd.to_numeric(imputed_imputed_cells.values.flatten(), errors='coerce')

    # Filter out any NaNs introduced by coercion or originally present
    valid_mask = ~np.isnan(original_flat) & ~np.isnan(imputed_flat)
    original_valid = original_flat[valid_mask]
    imputed_valid = imputed_flat[valid_mask]

    # Calculate MAE, handle case with no valid points
    if len(original_valid) == 0:
        return np.nan # Return NaN if no valid points to compare

    mae = np.mean(np.abs(original_valid - imputed_valid))
    return mae

# Apply the same logic to other evaluation functions that might fail on non-numeric data
def evaluate_imputation_rmse(original, imputed, mask):
    """Calculate Root Mean Squared Error only on imputed cells."""
    original_df = pd.DataFrame(original)
    imputed_df = pd.DataFrame(imputed)
    mask_df = pd.DataFrame(mask)

    original_imputed_cells = original_df[mask_df]
    imputed_imputed_cells = imputed_df[mask_df]

    original_flat = pd.to_numeric(original_imputed_cells.values.flatten(), errors='coerce')
    imputed_flat = pd.to_numeric(imputed_imputed_cells.values.flatten(), errors='coerce')

    valid_mask = ~np.isnan(original_flat) & ~np.isnan(imputed_flat)
    original_valid = original_flat[valid_mask]
    imputed_valid = imputed_flat[valid_mask]

    if len(original_valid) == 0:
        return np.nan

    rmse = np.sqrt(np.mean((original_valid - imputed_valid)**2))
    return rmse

def evaluate_imputation_mae_1(original, imputed, mask):
    """
    Calculates imputation accuracy for values that were originally missing.
    Focuses on numeric columns for difference calculation.
    Returns a measure like Mean Absolute Error (MAE) or a simple 'correct count'
    for categorical if a direct comparison is feasible.
    Here, we'll return MAE for numeric columns at imputed locations.
    """
    # Select only numeric columns
    numeric_cols = original.select_dtypes(include=np.number).columns.tolist()

    if not numeric_cols:
        print("Warning: No numeric columns found for Imputation Accuracy calculation.")
        return np.nan

    original_numeric = original[numeric_cols]
    imputed_numeric = imputed[numeric_cols]
    mask_numeric = mask[numeric_cols] # Apply mask to only numeric columns

    # Flatten and filter based on the mask of originally missing values in numeric columns
    original_flat = original_numeric[mask_numeric].values.flatten()
    imputed_flat = imputed_numeric[mask_numeric].values.flatten()

    # Filter out NaNs
    valid_mask = ~np.isnan(original_flat) & ~np.isnan(imputed_flat)
    original_valid = original_flat[valid_mask]
    imputed_valid = imputed_flat[valid_mask]

    if len(original_valid) == 0:
        print("Warning: No valid numeric data points found at missing locations for Imputation Accuracy calculation.")
        return np.nan # Avoid division by zero

    # Calculate MAE for numeric columns
    mae = np.mean(np.abs(original_valid - imputed_valid))
    return mae # Using MAE as a proxy for "accuracy" on numeric imputations

# RMSE / MSE – measures the accuracy of imputation (especially for numeric data, or ordinal-encoded categorical data).

#2. Model Performance (F1, Accuracy) – how well an ML model performs on downstream tasks (e.g., classification) using the imputed data.
def evaluate_model_performance(imputed_data, labels):
    """Evaluate downstream model performance"""
    encoder = OrdinalEncoder()
    X_encoded = encoder.fit_transform(imputed_data)
    X_train, X_test, y_train, y_test = train_test_split(X_encoded, labels, test_size=0.2, random_state=42)

    clf = RandomForestClassifier(random_state=42)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)



    joblib.dump(encoder, "ordinal_encoder.pkl")
    print("✅ Encoder saved as 'ordinal_encoder.pkl'")


    # ✅ Save the trained model
    joblib.dump(clf, "SHF_MVI_model.pkl")
    print("✅ Model saved successfully as 'SHF_MVI_model.pkl'")


    return accuracy_score(y_test, y_pred), f1_score(y_test, y_pred, average='weighted')


def evaluate_precision(imputed_data, true_labels):
    """Calculate precision score for imputed data with robust error handling"""
    try:
        encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
        X_encoded = encoder.fit_transform(imputed_data)

        X_train, X_test, y_train, y_test = train_test_split(
            X_encoded, true_labels, test_size=0.2, random_state=42, stratify=true_labels
        )

        clf = RandomForestClassifier(random_state=42)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)

        return precision_score(y_test, y_pred, average='weighted')
    except Exception as e:
        print(f"Precision evaluation error: {str(e)}")
        return 0

from scipy import stats
import numpy as np



def calculate_distribution_pvalue(sample):
    """
    Calculates a p-value representing how much the sample distribution
    deviates from a normal distribution using the Kolmogorov-Smirnov test.
    Returns a smaller p-value for greater deviation.
    Assumes input 'sample' is a pandas Series of numeric data with no NaNs.
    """
    # Ensure input is a Series and contains numeric data
    if not isinstance(sample, pd.Series):
        try:
            sample = pd.Series(sample)
        except Exception:
            print("Warning: calculate_distribution_pvalue received non-Series input.")
            return np.nan # Return NaN if conversion fails

    # Drop NaNs (though the caller should ideally handle this)
    sample = sample.dropna()

    # Need at least 2 non-NaN values to calculate std deviation
    if len(sample) < 2:
        return np.nan # Not enough data to perform test

    # Check if all values are the same (std dev will be 0)
    if sample.nunique() == 1:
        return 1.0 # Perfect 'distribution' (a single point), doesn't deviate from itself


    try:
        # Standardize the sample
        # Add a small epsilon to std() to prevent division by zero if sample.std() is exactly 0
        std_dev = sample.std()
        if std_dev < 1e-9: # Check against a small threshold instead of exactly 0
             return 1.0 # Very low variance, considered close to normal

        standardized = (sample - sample.mean()) / std_dev

        # Use Kolmogorov–Smirnov with standard normal as reference
        _, p = stats.kstest(standardized, 'norm')
        return p
    except Exception as e:
        print(f"Error in calculate_distribution_pvalue: {e}")
        return 0 # Return NaN if KS test fails for other reasons




def calculate_distribution_similarity(original, imputed,mask):
    """Calculate similarity of value distributions between original and imputed data"""
    from scipy.stats import wasserstein_distance

    similarities = []
    for col in original.columns:
        orig_counts = original[col].value_counts(normalize=True)
        imp_counts = imputed[col].value_counts(normalize=True)

        # Align indices
        all_categories = list(set(orig_counts.index) | set(imp_counts.index))
        orig_dist = [orig_counts.get(cat, 0) for cat in all_categories]
        imp_dist = [imp_counts.get(cat, 0) for cat in all_categories]

        similarities.append(1 - wasserstein_distance(orig_dist, imp_dist))

    return np.mean(similarities)

def calculate_correlation_preservation(original, imputed):
    """Measure how well correlations are preserved after imputation"""
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    orig_encoded = encoder.fit_transform(original)
    imp_encoded = encoder.transform(imputed)

    orig_corr = np.corrcoef(orig_encoded.T)
    imp_corr = np.corrcoef(imp_encoded.T)

    # Calculate correlation matrix difference
    return np.linalg.norm(orig_corr - imp_corr)

def evaluate_imputation(original_df, imputed_df, mask,
                        labels,missing_rate,
                        tech_name):
    results = {}
    original_df = pd.DataFrame(original_df).copy()
    imputed_df = pd.DataFrame(imputed_df).copy()
    ##########mask = pd.DataFrame(mask).copy()
    assert mask.shape == original_df.shape
    assert (mask.columns == original_df.columns).all()
    assert (mask.index == original_df.index).all()


    imputed_df = imputed_df.reindex(columns=original_df.columns)
    # Basic accuracy metrics
    results['Technique Name'] = tech_name
    results['accuracy'] = evaluate_imputation_accuracy(original_df, imputed_df, mask)
    results['rmse'] = evaluate_imputation_rmse(original_df, imputed_df, mask)
    results['mae'] = evaluate_imputation_mae(original_df, imputed_df, mask)

    # Downstream model performance
    acc, f1 = evaluate_model_performance(imputed_df, labels)
    results['model_accuracy'] = acc
    results['model_f1'] = f1
    results['precision'] = evaluate_precision(imputed_df, labels)
    results['missing_rate'] =missing_rate

    mae = mean_absolute_error(original_df, imputed_df)
    mse = mean_squared_error(original_df, imputed_df)
    rmse = np.sqrt(mse)
    r2 = r2_score(original_df, imputed_df)

    print(f"MAE: {mae}, RMSE: {rmse}, R²: {r2}")
    results['mean_absolute_error'] = mae
    results['mean_squared_error'] = mse
    results['rmsermse'] = rmse
    results['r2_score'] =r2


    try:
        results['distribution_similarity'] = calculate_distribution_similarity(original_df, imputed_df, mask)
    except Exception as e:
        print(f"Error calculating distribution similarity: {e}")
        results['distribution_similarity'] = 0
    try:
        results['correlation_preservation'] = calculate_correlation_preservation(original_df, imputed_df)
    except Exception as e:
        print(f"Error calculating correlation preservation: {e}")
        results['correlation_preservation'] = 0

    return results
```

```python
def impute_gan(df_missing, epochs=100, batch_size=32):
    """GAN-based imputation for missing data"""
    # Encode data
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Normalize
    min_val = np.nanmin(encoded)
    max_val = np.nanmax(encoded)
    encoded_normalized = (encoded - min_val) / (max_val - min_val)

    # Create mask for missing values
    mask = np.isnan(encoded_normalized)
    X_train = np.nan_to_num(encoded_normalized, nan=0)

    # GAN components
    latent_dim = encoded_normalized.shape[1]

    # Generator
    generator = Sequential([
        Dense(64, activation='relu', input_dim=latent_dim),
        Dense(128, activation='relu'),
        Dense(latent_dim, activation='sigmoid')
    ])

    # Discriminator
    discriminator = Sequential([
        Dense(128, activation='relu', input_dim=latent_dim),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid')
    ])

    # Combined GAN
    discriminator.compile(optimizer=Adam(0.0002, 0.5), loss='binary_crossentropy')
    discriminator.trainable = False

    gan_input = Input(shape=(latent_dim,))
    fake_data = generator(gan_input)
    gan_output = discriminator(fake_data)

    gan = Model(gan_input, gan_output)
    gan.compile(optimizer=Adam(0.0002, 0.5), loss='binary_crossentropy')

    # Training
    valid = np.ones((batch_size, 1))
    fake = np.zeros((batch_size, 1))

    for epoch in range(epochs):
        # Train discriminator
        idx = np.random.randint(0, X_train.shape[0], batch_size)
        real_data = X_train[idx]

        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        gen_data = generator.predict(noise, verbose=0)

        d_loss_real = discriminator.train_on_batch(real_data, valid)
        d_loss_fake = discriminator.train_on_batch(gen_data, fake)
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

        # Train generator
        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        g_loss = gan.train_on_batch(noise, valid)

    # Impute missing values
    noise = np.random.normal(0, 1, (X_train.shape[0], latent_dim))
    imputed_encoded = generator.predict(noise, verbose=0)

    # Only replace missing values
    imputed_encoded = np.where(mask, imputed_encoded, encoded_normalized)

    # Denormalize and convert back
    imputed_encoded = imputed_encoded * (max_val - min_val) + min_val
    imputed_encoded = np.round(imputed_encoded).astype(int)
    imputed_encoded = np.clip(imputed_encoded, 0, len(encoder.categories_[0])-1)
    return pd.DataFrame(encoder.inverse_transform(imputed_encoded),
                       columns=df_missing.columns)


def gain_impute(df_missing, batch_size=64, epochs=100):
    """GAIN (Generative Adversarial Imputation Network) implementation."""
    try:
        # 1. Data preparation
        data = df_missing.values
        mask = ~np.isnan(data)
        data[mask == False] = 0  # Temporarily fill missing with 0

        # Normalize data
        data_min = np.nanmin(data, axis=0)
        data_max = np.nanmax(data, axis=0)
        data_norm = (data - data_min) / (data_max - data_min + 1e-8)

        # 2. Generator model
        def build_generator():
            inputs = Input(shape=(data.shape[1],))
            x = Dense(128, activation='relu')(inputs)
            x = Dense(128, activation='relu')(x)
            outputs = Dense(data.shape[1], activation='sigmoid')(x)
            return Model(inputs, outputs)

        # 3. Discriminator model
        def build_discriminator():
            inputs = Input(shape=(data.shape[1],))
            x = Dense(128, activation='relu')(inputs)
            x = Dense(128, activation='relu')(x)
            outputs = Dense(data.shape[1], activation='sigmoid')(x)
            return Model(inputs, outputs)

        # 4. Build and compile models
        generator = build_generator()
        discriminator = build_discriminator()

        # Use tf.function for proper tensor handling
        @tf.function
        def train_step(batch_data, batch_mask):
            with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
                # Generator forward pass
                gen_output = generator(batch_data, training=True)

                # Combine real and generated data
                combined_data = batch_data * batch_mask + gen_output * (1 - batch_mask)

                # Discriminator predictions
                real_output = discriminator(batch_data, training=True)
                fake_output = discriminator(combined_data, training=True)

                # Loss calculations
                gen_loss = -tf.reduce_mean(fake_output)
                disc_loss = tf.reduce_mean(fake_output) - tf.reduce_mean(real_output)

            # Apply gradients
            gen_gradients = gen_tape.gradient(gen_loss, generator.trainable_variables)
            disc_gradients = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

            return gen_loss, disc_loss

        # 5. Training loop
        optimizer = Adam(0.0001)
        generator_optimizer = optimizer
        discriminator_optimizer = optimizer

        for epoch in range(epochs):
            for i in range(0, len(data), batch_size):
                batch_data = data_norm[i:i + batch_size]
                batch_mask = mask[i:i + batch_size].astype(float)

                gen_loss, disc_loss = train_step(
                    tf.convert_to_tensor(batch_data, dtype=tf.float32),
                    tf.convert_to_tensor(batch_mask, dtype=tf.float32)
                )

        # 6. Imputation
        gen_imputations = generator.predict(data_norm)
        imputed_data = data_norm * mask + gen_imputations * (1 - mask)
        imputed_data = imputed_data * (data_max - data_min) + data_min

        return pd.DataFrame(imputed_data, columns=df_missing.columns)

    except Exception as e:
        print(f"GAIN failed: {str(e)}")
        print("Falling back to iterative imputer...")
        from sklearn.impute import IterativeImputer
        return pd.DataFrame(
            IterativeImputer().fit_transform(df_missing),
            columns=df_missing.columns
        )




def impute_gain(df_missing, batch_size=32, epochs=100, hint_rate=0.9, alpha=10):
    """
    GAIN (Generative Adversarial Imputation Nets) imputation for categorical data.
    """
    # Encode categorical data
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Normalize data
    # Handle cases where columns might be all NaN before normalization
    min_val = np.nanmin(encoded, axis=0)
    max_val = np.nanmax(encoded, axis=0)

    # Replace NaN mins/maxs with a default range if the column was all NaN
    min_val[np.isnan(min_val)] = 0
    max_val[np.isnan(max_val)] = 1 # Use a default range 0-1 if column was empty

    # Add a small epsilon to avoid division by zero for columns with constant values
    range_vals = max_val - min_val + K.epsilon()

    X_normalized_original = (encoded - min_val) / range_vals

    # Create mask (1: observed, 0: missing)
    mask_np = (~np.isnan(encoded)).astype(float)
    X_normalized_original_filled = np.nan_to_num(X_normalized_original, nan=0)


    # Data dimensions
    data_dim = X_normalized_original_filled.shape[1]
    z_dim = data_dim # Latent dim often same as data dim in GAIN variations

    # Use tf.keras.layers.Concatenate within the model definition
    from tensorflow.keras.layers import Concatenate

    # Generator network
    def build_generator():
        # Input is concatenated [Z, M]
        input_layer = Input(shape=(data_dim * 2,))
        x = Dense(256, activation='relu')(input_layer)
        x = BatchNormalization()(x)
        x = Dense(256, activation='relu')(x)
        x = BatchNormalization()(x)
        outputs = Dense(data_dim, activation='sigmoid')(x) # Output is X_hat (imputed values for all features)
        return Model(inputs=input_layer, outputs=outputs, name='generator')

    # Discriminator network
    def build_discriminator():
        # Input is concatenated [X_tilde, H]
        input_layer = Input(shape=(data_dim * 2,))
        x = Dense(256, activation='relu')(input_layer)
        x = BatchNormalization()(x)
        x = Dense(256, activation='relu')(x)
        x = BatchNormalization()(x)
        outputs = Dense(1, activation='sigmoid')(x) # Single output: Probability of being 'real'
        return Model(inputs=input_layer, outputs=outputs, name='discriminator')

    # Build models
    generator = build_generator()
    discriminator = build_discriminator()

    # Optimizers
    generator_optimizer = Adam(learning_rate=0.0001)
    discriminator_optimizer = Adam(learning_rate=0.0001)

    # Use tf.GradientTape for the custom training loop
    @tf.function
    def train_step(X_batch, M_batch):
        batch_size_actual = tf.shape(X_batch)[0]

        # Generate Z (input for generator) - Typically X_batch with missing filled with random noise
        Z_batch = tf.identity(X_batch) # Start with the batch data (normalized, NaNs filled with 0)
        # Add random noise specifically where values are missing
        random_noise = tf.random.uniform(tf.shape(Z_batch), 0, 1)
        Z_batch = tf.where(M_batch == 0, random_noise, Z_batch) # Replace missing locations with noise

        # Create Hint vector (H)
        random_mask = tf.random.uniform(tf.shape(M_batch))
        H_batch = tf.where(random_mask < hint_rate, M_batch, tf.zeros_like(M_batch))

        # --- Train Discriminator ---
        with tf.GradientTape() as disc_tape:
            # Input to generator: [Z, M]
            gen_input_disc = tf.concat([Z_batch, M_batch], axis=1)
            # Generator output (X_hat)
            X_hat_batch = generator(gen_input_disc, training=True)
            # Combined data (X_tilde): M * X_batch + (1 - M) * X_hat_batch
            X_tilde_batch = M_batch * X_batch + (1 - M_batch) * X_hat_batch

            # Input to discriminator: [X_real, H] and [X_tilde, H]
            disc_input_real = tf.concat([X_batch, H_batch], axis=1) # Discriminator sees real data + hint
            disc_input_fake = tf.concat([X_tilde_batch, H_batch], axis=1) # Discriminator sees combined data + hint

            # Discriminator predictions
            real_output = discriminator(disc_input_real, training=True)
            fake_output = discriminator(disc_input_fake, training=True)

            # Discriminator loss (standard GAN: BCE)
            # Ensure target labels have the same dtype as predictions
            real_labels = tf.ones_like(real_output, dtype=tf.float32)
            fake_labels = tf.zeros_like(fake_output, dtype=tf.float32)

            real_loss = binary_crossentropy(real_labels, real_output)
            fake_loss = binary_crossentropy(fake_labels, fake_output)
            disc_loss = tf.reduce_mean(real_loss) + tf.reduce_mean(fake_loss)

        # Apply discriminator gradients
        disc_gradients = disc_tape.gradient(disc_loss, discriminator.trainable_variables)
        discriminator_optimizer.apply_gradients(zip(disc_gradients, discriminator.trainable_variables))

        # --- Train Generator ---
        with tf.GradientTape() as gen_tape:
             # Input to generator: [Z, M] (same as for discriminator training step)
            gen_input_gen = tf.concat([Z_batch, M_batch], axis=1)
            # Generator output (X_hat)
            X_hat_batch = generator(gen_input_gen, training=True) # Generator is trained here
            # Combined data (X_tilde) - used for reconstruction loss and GAN loss
            X_tilde_batch = M_batch * X_batch + (1 - M_batch) * X_hat_batch

            # Input to discriminator for generator training: [X_tilde, H]
            # The discriminator is NOT trained in this step, so training=False
            disc_input_fake_gen = tf.concat([X_tilde_batch, H_batch], axis=1)
            fake_output_gen = discriminator(disc_input_fake_gen, training=False) # Discriminator weights are frozen

            # Generator loss (GAN + Reconstruction)
            # GAN loss: Generator wants discriminator to output 1 (real) for the generated data
            gen_gan_loss = binary_crossentropy(tf.ones_like(fake_output_gen, dtype=tf.float32), fake_output_gen)

            # Reconstruction loss: How close are the imputed values (X_hat) to the original observed values (X_batch)?
            # Only calculated for the observed parts (where M=1)
            # tf.abs handles differences correctly. K.sum and K.epsilon are fine here.
            reconstruction_loss = K.sum(K.abs((X_batch - X_hat_batch) * M_batch)) / (K.sum(M_batch) + K.epsilon())

            gen_loss = tf.reduce_mean(gen_gan_loss) + alpha * reconstruction_loss # Total generator loss

        # Apply generator gradients
        gen_gradients = gen_tape.gradient(gen_loss, generator.trainable_variables)
        generator_optimizer.apply_gradients(zip(gen_gradients, generator.trainable_variables))

        return disc_loss, gen_loss, reconstruction_loss # Return reconstruction loss for monitoring

    # Training loop
    n_batches = int(np.ceil(X_normalized_original_filled.shape[0] / batch_size))

    # Convert full dataset to tensors once
    X_full_tf = tf.convert_to_tensor(X_normalized_original_filled, dtype=tf.float32)
    mask_full_tf = tf.convert_to_tensor(mask_np, dtype=tf.float32)


    for epoch in range(epochs):
        epoch_d_loss = 0.0
        epoch_g_loss = 0.0
        epoch_recon_loss = 0.0

        for batch_idx in range(n_batches):
            # Get batch data using slicing on the full tensors
            start = batch_idx * batch_size
            end = min((batch_idx + 1) * batch_size, X_full_tf.shape[0])
            X_batch_tf = X_full_tf[start:end]
            M_batch_tf = mask_full_tf[start:end]

            d_loss, g_loss, recon_loss = train_step(X_batch_tf, M_batch_tf)

            epoch_d_loss += d_loss
            epoch_g_loss += g_loss
            epoch_recon_loss += recon_loss


        # Calculate average losses for the epoch
        avg_d_loss = epoch_d_loss / n_batches
        avg_g_loss = epoch_g_loss / n_batches
        avg_recon_loss = epoch_recon_loss / n_batches


        # Print progress periodically
        if (epoch + 1) % 10 == 0 or epoch == 0:
             print(f"Epoch {epoch+1}/{epochs} - D loss: {avg_d_loss:.4f}, G loss: {avg_g_loss:.4f}, Recon loss: {avg_recon_loss:.4f}")



    Z_impute_tf = tf.convert_to_tensor(np.nan_to_num(X_normalized_original, nan=0), dtype=tf.float32)
    mask_impute_tf = tf.convert_to_tensor(mask_np, dtype=tf.float32)

    gen_input_impute_final = tf.concat([Z_impute_tf, mask_impute_tf], axis=1)

    # Get generator's prediction for the whole dataset
    X_hat_final_tf = generator(gen_input_impute_final, training=False) # Predict using trained generator

    # Convert back to numpy
    X_hat_final_np = X_hat_final_tf.numpy()

    # Combine observed data (from original normalized) with generator's output for missing parts
    X_imputed_normalized = X_normalized_original.copy() # Start with the original normalized data (with NaNs)
    # Fill the NaN locations with the generator's predictions
    # Use mask_np for indexing
    missing_locations = (mask_np == 0)
    X_imputed_normalized[missing_locations] = X_hat_final_np[missing_locations]

    # Denormalize and convert back to categories
    # Use the saved min_val and range_vals (max_val - min_val + epsilon) for denormalization
    X_imputed_denormalized = X_imputed_normalized * range_vals + min_val

    # Round and convert to integer indices
    X_imputed_rounded = np.round(X_imputed_denormalized)
    X_imputed_int = X_imputed_rounded.astype(int)

    # Clip to valid category ranges based on the original encoder categories
    for i in range(X_imputed_int.shape[1]):
        if i < len(encoder.categories_):
             num_categories = len(encoder.categories_[i])
             if num_categories > 0:
                X_imputed_int[:, i] = np.clip(
                    X_imputed_int[:, i],
                    0,
                    num_categories - 1
                )
             else:
                 # If original column was encoded but had no unique categories (e.g., all NaNs),
                 # clip to 0 or handle as appropriate. Setting to 0 might be okay if it maps to a valid category index 0.
                 print(f"Warning: Column {i} had no unique categories in original data. Clipping to 0.")
                 X_imputed_int[:, i] = 0
        else:
             # This case should ideally not happen if column count matches
             print(f"Error: Column index {i} out of bounds for encoder categories during clipping.")
             X_imputed_int[:, i] = 0 # Default to 0 as a safe fallback

    # Use inverse_transform on the final integer array
    imputed_data_inverse_transformed = encoder.inverse_transform(X_imputed_int)

    # Clean up TensorFlow session and large variables
    K.clear_session()
    del generator, discriminator, generator_optimizer, discriminator_optimizer
    # Delete tensors if they are large
    del X_full_tf, mask_full_tf, Z_impute_tf, mask_impute_tf, gen_input_impute_final, X_hat_final_tf
    gc.collect()
    # tf.compat.v1.reset_default_graph() # Only needed for older TF versions or specific contexts


    return pd.DataFrame(imputed_data_inverse_transformed, columns=df_missing.columns)

def impute_gain_(df_missing, batch_size=32, epochs=100, hint_rate=0.9, alpha=10):
    """
    GAIN (Generative Adversarial Imputation Nets) imputation for categorical data.
    """
    # Encode categorical data
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Normalize data
    min_val = np.nanmin(encoded)
    max_val = np.nanmax(encoded)
    X = (encoded - min_val) / (max_val - min_val)

    # Create mask (1: observed, 0: missing)
    mask = (~np.isnan(X)).astype(float)
    X = np.nan_to_num(X, nan=0)

    # Data dimensions
    data_dim = X.shape[1]
    z_dim = data_dim
    def build_generator():
        inputs = Input(shape=(data_dim * 2,))  # Concatenated data + mask

        # Revised generator input to be the already concatenated tensor
        input_layer = Input(shape=(data_dim * 2,))
        x = Dense(256, activation='relu')(input_layer)
        x = BatchNormalization()(x)
        x = Dense(256, activation='relu')(x)
        x = BatchNormalization()(x)
        outputs = Dense(data_dim, activation='sigmoid')(x)
        return Model(inputs=input_layer, outputs=outputs)

    def build_discriminator():
        inputs = Input(shape=(data_dim * 2,))  # Concatenated data + hint
        x = Dense(256, activation='relu')(inputs)
        x = BatchNormalization()(x)
        x = Dense(256, activation='relu')(x)
        x = BatchNormalization()(x)
        outputs = Dense(data_dim, activation='sigmoid')(x)
        return Model(inputs=inputs, outputs=outputs)

    # Generator network
    def build_generator_():
        inputs = Input(shape=(data_dim * 2,))  # Concatenated data + mask
        x = Dense(256, activation='relu')(inputs)
        x = BatchNormalization()(x)
        x = Dense(256, activation='relu')(x)
        x = BatchNormalization()(x)
        outputs = Dense(data_dim, activation='sigmoid')(x)
        return Model(inputs, outputs)

    # Discriminator network
    def build_discriminator_():
        inputs = Input(shape=(data_dim * 2,))  # Concatenated data + hint
        x = Dense(256, activation='relu')(inputs)
        x = BatchNormalization()(x)
        x = Dense(256, activation='relu')(x)
        x = BatchNormalization()(x)
        outputs = Dense(data_dim, activation='sigmoid')(x)
        return Model(inputs, outputs)

    # Build and compile models
    generator = build_generator()
    discriminator = build_discriminator()

    # Combined model
    z = Input(shape=(data_dim,))
    mask_input = Input(shape=(data_dim,))
    concatenated_input_for_generator = Concatenate(axis=1)([z, mask_input])
    x_hat = generator(concatenated_input_for_generator)

    #x_hat = generator(K.concatenate([z, mask_input], axis=1))
    x_tilde = mask_input * z + (1 - mask_input) * x_hat

    # For the model definition, we need the placeholder for the hint input to the discriminator
    hint_input = Input(shape=(data_dim,))

    # Use the Concatenate layer for discriminator input as well
    concatenated_input_for_discriminator = Concatenate(axis=1)([x_tilde, hint_input])
    d_logits = discriminator(concatenated_input_for_discriminator)

    # Define the combined GAN model
    # The combined model needs all the primary inputs
    gan = Model(inputs=[z, mask_input, hint_input], outputs=d_logits)

    # Define loss functions
    def generator_loss(y_true, y_pred):
        # MSE loss
        mse_loss = K.mean((y_true - y_pred) ** 2, axis=1)
        # Adversarial loss
        adv_loss = K.mean(binary_crossentropy(y_true, y_pred), axis=1) # This binary_crossentropy use might be incorrect depending on GAN objective


    # Compile models
    generator.compile(optimizer='adam', loss=generator_loss)
    discriminator.compile(optimizer='adam', loss='binary_crossentropy')

    # Training loop
    n_batches = int(np.ceil(X.shape[0] / batch_size))

    for epoch in range(epochs):
        for batch_idx in range(n_batches):
            # Get batch
            start = batch_idx * batch_size
            end = min((batch_idx + 1) * batch_size, X.shape[0])
            X_batch = X[start:end]
            M_batch = mask[start:end]

            # Generate fake samples
            Z_batch = X_batch.copy()
            Z_batch[M_batch == 0] = np.random.uniform(0, 1, size=np.sum(M_batch == 0))

            # Train discriminator
            X_hat = generator.predict(K.concatenate([Z_batch, M_batch], axis=1), verbose=0)
            X_tilde = M_batch * X_batch + (1 - M_batch) * X_hat

            # Create hint vectors
            H_batch = np.where(np.random.uniform(size=M_batch.shape) < hint_rate,
                             M_batch,
                             np.zeros_like(M_batch))

            # Labels (1 for real, 0 for fake)
            D_labels = M_batch.copy()

            d_loss = discriminator.train_on_batch(
                K.concatenate([X_tilde, H_batch], axis=1),
                D_labels
            )

            # Train generator
            g_loss = generator.train_on_batch(
                K.concatenate([Z_batch, M_batch], axis=1),
                X_batch
            )

    # Impute missing values
    Z = X.copy()
    Z[mask == 0] = np.random.uniform(0, 1, size=np.sum(mask == 0))
    X_hat = generator.predict(K.concatenate([Z, mask], axis=1), verbose=0)
    X_imputed = mask * X + (1 - mask) * X_hat

    # Denormalize and convert back to categories
    X_imputed = X_imputed * (max_val - min_val) + min_val
    X_imputed = np.round(X_imputed).astype(int)
    X_imputed = np.clip(X_imputed, 0, len(encoder.categories_[0])-1)

    return pd.DataFrame(encoder.inverse_transform(X_imputed), columns=df_missing.columns)

def impute_vae(df_missing, batch_size=32, epochs=50, latent_dim=16, beta=1.0):
    """VAE-based imputation for missing data."""
    print("Starting VAE imputation...")
    try:
        # 1. Data preparation
        # Encode categorical data
        encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
        encoded = encoder.fit_transform(df_missing)

        # Normalize data
        col_mins = np.nanmin(encoded, axis=0)
        col_maxs = np.nanmax(encoded, axis=0)

        # Handle columns with all NaN or constant values
        range_vals = col_maxs - col_mins
        range_vals[range_vals == 0] = 1e-8 # Avoid division by zero
        col_mins[np.isnan(col_mins)] = 0
        col_maxs[np.isnan(col_maxs)] = 1 # Default range

        X_normalized_original = (encoded - col_mins) / (range_vals)

        # Create mask (1: observed, 0: missing)
        mask_np = (~np.isnan(encoded)).astype(float)
        X_normalized_original_filled = np.nan_to_num(X_normalized_original, nan=0.5) # Fill NaNs with a neutral value like 0.5

        X_train = X_normalized_original_filled
        M_train = mask_np

        original_dim = X_train.shape[1]

        # 2. VAE Architecture
        # Encoder
        inputs = Input(shape=(original_dim,))
        # Combine input data with mask for encoder
        encoder_input = Concatenate()([inputs, Input(shape=(original_dim,))]) # Placeholder for mask input

        x = Dense(128, activation='relu')(encoder_input)
        x = BatchNormalization()(x)
        x = Dense(64, activation='relu')(x)
        x = BatchNormalization()(x)

        z_mean = Dense(latent_dim, name='z_mean')(x)
        z_log_var = Dense(latent_dim, name='z_log_var')(x)

        # Reparameterization trick
        def sampling(args):
            z_mean, z_log_var = args
            epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent_dim), mean=0., stddev=1.)
            return z_mean + K.exp(0.5 * z_log_var) * epsilon

        z = Lambda(sampling, output_shape=(latent_dim,), name='z')([z_mean, z_log_var])

        encoder = Model([inputs, Input(shape=(original_dim,))], [z_mean, z_log_var, z], name='encoder')

        # Decoder
        latent_inputs = Input(shape=(latent_dim,))
        decoder_input = Concatenate()([latent_inputs, Input(shape=(original_dim,))]) # Placeholder for mask input

        x = Dense(64, activation='relu')(decoder_input)
        x = BatchNormalization()(x)
        x = Dense(128, activation='relu')(x)
        x = BatchNormalization()(x)
        outputs = Dense(original_dim, activation='sigmoid')(x) # Output values are normalized 0-1

        decoder = Model([latent_inputs, Input(shape=(original_dim,))], outputs, name='decoder')

        # VAE (combined model)
        vae_input = Input(shape=(original_dim,))
        mask_input = Input(shape=(original_dim,)) # Explicit mask input

        # Pass input data AND mask to encoder
        z_mean_vae, z_log_var_vae, z_vae = encoder([vae_input, mask_input])

        # Pass latent representation AND mask to decoder
        vae_output = decoder([z_vae, mask_input])

        # Only the combined model needs to compute the full VAE loss
        vae = Model([vae_input, mask_input], vae_output, name='vae_imputer')


        # 3. Custom VAE Loss Layer
        class VAELossLayer(tf.keras.layers.Layer):
            def __init__(self, beta=1.0, **kwargs):
                super(VAELossLayer, self).__init__(**kwargs)
                self.beta = beta
                self.binary_crossentropy = tf.keras.losses.BinaryCrossentropy(from_logits=False)

            def call(self, inputs):
                # inputs = [true_x, reconstructed_x, z_mean, z_log_var]
                true_x, reconstructed_x, z_mean, z_log_var = inputs

                reconstruction_loss_per_feature = self.binary_crossentropy(true_x, reconstructed_x) * M_train # Apply mask directly here if M_train has batch dim
                reconstruction_loss = tf.reduce_sum(reconstruction_loss_per_feature, axis=-1) # Sum loss per feature


                # KL Divergence loss
                kl_loss = -0.5 * tf.reduce_sum(1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var), axis=-1)


                # Total VAE loss: Reconstruction Loss + beta * KL Divergence
                total_vae_loss = tf.reduce_mean(reconstruction_loss + self.beta * kl_loss)

                self.add_loss(total_vae_loss)

                # Return the reconstructed output, which is the model's final prediction
                return reconstructed_x

        # Add the custom loss layer to the VAE model
        # This layer takes the necessary tensors as input and adds the loss to the model
        vae_output_with_loss = VAELossLayer(beta=beta)([vae_input, vae_output, z_mean_vae, z_log_var_vae])

        # Create the final model that outputs the imputed data
        vae_imputer = Model(inputs=[vae_input, mask_input], outputs=vae_output_with_loss, name='vae_final')

        # 4. Compile the model
        # The loss is added via add_loss(), so we can compile with a dummy loss or None
        # Using a dummy loss like 'mse' on the output, but the primary training signal comes from the custom loss layer
        vae_imputer.compile(optimizer=Adam(learning_rate=0.001), loss='mse') # Use mse for the final output check if needed, but VAE loss is main driver


        # 5. Prepare data for training (Input should be [data, mask])
        train_dataset = tf.data.Dataset.from_tensor_slices(((X_train, M_train), X_train)).batch(batch_size).shuffle(1024)

        # 6. Train the VAE
        print("Training VAE...")
        vae_imputer.fit(train_dataset, epochs=epochs, verbose=verbose)

        # 7. Impute missing values
        print("Imputing missing values...")
        # Predict using the model, providing both the data (with NaNs filled) and the mask
        predicted_normalized = vae_imputer.predict([X_train, M_train], batch_size=batch_size)

        # Only replace missing values (where mask is 0) with the predicted values
        imputed_normalized = X_normalized_original_filled * M_train + predicted_normalized * (1 - M_train)


        # 8. Denormalize and convert back
        imputed_encoded = imputed_normalized * (range_vals) + col_mins

        # Round to nearest integer for categorical-like data (if applicable)
        # This step depends heavily on whether the original data was truly categorical or continuous
        # If it's mixed, this needs more sophistication. Assuming encoded values represent categories.
        imputed_encoded = np.round(imputed_encoded)

        # Clip to valid category ranges based on the original encoder categories
        # This requires accessing encoder.categories_
        for i in range(imputed_encoded.shape[1]):
             # Ensure the index is within the bounds of encoder.categories_
            if i < len(encoder.categories_):
                valid_range_min = -1 if encoder.handle_unknown == 'use_encoded_value' else 0
                valid_range_max = len(encoder.categories_[i]) - 1
                imputed_encoded[:, i] = np.clip(imputed_encoded[:, i], valid_range_min, valid_range_max).astype(int)
            else:
                 # Handle columns that might not have been considered categorical by encoder
                 # or were added later. A simple clip to non-negative might be needed or more complex logic.
                 print(f"Warning: Column {i} not found in encoder categories. Clipping to >= 0.")
                 imputed_encoded[:, i] = np.clip(imputed_encoded[:, i], 0, None).astype(int) # Default clip


        # Inverse transform to get original data types/values
        # Handle the case where inverse_transform might fail if imputed values are outside original categories
        try:
            imputed_df = pd.DataFrame(encoder.inverse_transform(imputed_encoded), columns=df_missing.columns)
        except Exception as e:
             print(f"Error during inverse transform: {e}. Check imputed_encoded values.")
             # Fallback or more robust handling needed here
             imputed_df = df_missing.copy() # Return original or partially imputed
             # Optionally, try inverse transforming column by column if needed

        print("VAE imputation completed.")
        return imputed_df

    except Exception as e:
        print(f"Error in VAE imputation: {str(e)}")
        import traceback
        traceback.print_exc() # Print full traceback for debugging
        print("Falling back to IterativeImputer...")
        # Fallback to a more robust method
        from sklearn.impute import IterativeImputer
        # Ensure IterativeImputer can handle mixed types or fall back further
        try:
             imputer = IterativeImputer(random_state=42)
             imputed_data = imputer.fit_transform(df_missing)
             return pd.DataFrame(imputed_data, columns=df_missing.columns)
        except Exception as fallback_e:
             print(f"IterativeImputer fallback also failed: {str(fallback_e)}")
             print("Falling back to mode imputation as a last resort.")
             return impute_mode(df_missing) # Last resort fallback

def impute_vae_(df_missing, batch_size=32, epochs=50, latent_dim=8, beta=1.0):
    """VAE-based imputation with revised loss calculation."""
    # Encode categorical data
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Handle columns that are all NaN after encoding
    col_mins = np.nanmin(encoded, axis=0)
    col_maxs = np.nanmax(encoded, axis=0)

    # Identify columns that were all NaN in the original data
    all_nan_cols_mask = np.all(np.isnan(encoded), axis=0)

    # Replace NaN mins/maxs with a default range if the column was all NaN
    min_val = np.where(all_nan_cols_mask, 0, col_mins)
    max_val = np.where(all_nan_cols_mask, 1, col_maxs) # Use a default range 0-1

    # Add a small epsilon to avoid division by zero for columns with constant values
    range_vals = max_val - min_val
    # Replace 0 range with 1 to avoid division by zero, affecting only columns with constant values > 0
    range_vals = np.where(range_vals == 0, 1e-10, range_vals)


    X_normalized_original = (encoded - min_val) / range_vals

    # Create mask (1: observed, 0: missing)
    mask_np = (~np.isnan(encoded)).astype(float)
    X_normalized_original_filled = np.nan_to_num(X_normalized_original, nan=0)

    # Convert to TensorFlow tensors
    X = tf.constant(X_normalized_original_filled, dtype=tf.float32)
    mask = tf.constant(mask_np, dtype=tf.float32)

    original_dim = X.shape[1] # Use X.shape[1] after converting to tensor

    # VAE Model definition
    # Define sampler as a Lambda layer
    def sampling(args):
        z_mean, z_log_var = args
        epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent_dim), mean=0., stddev=1.0)
        return z_mean + K.exp(0.5 * z_log_var) * epsilon

    # Build the VAE model
    def build_vae(original_dim, latent_dim):
        # Encoder
        inputs = Input(shape=(original_dim,))
        x = Dense(128, activation='relu')(inputs)
        x = BatchNormalization()(x)
        z_mean = Dense(latent_dim, name='z_mean')(x)
        z_log_var = Dense(latent_dim, name='z_log_var')(x)

        # Sampler layer
        z = Lambda(sampling, output_shape=(latent_dim,), name='z')([z_mean, z_log_var])

        # Decoder
        decoder_input = Input(shape=(latent_dim,))
        x_decoded = Dense(128, activation='relu')(decoder_input)
        x_decoded = BatchNormalization()(x_decoded)
        outputs = Dense(original_dim, activation='sigmoid')(x_decoded)
        decoder = Model(decoder_input, outputs, name='decoder')

        # VAE model (Encoder + Decoder)
        x_decoded_mean = decoder(z)

        # Define VAE loss inside the function where z_mean and z_log_var are KerasTensors
        def vae_loss(y_true, y_pred): # y_true is the original data, y_pred is the reconstruction
            # Reconstruction loss (use masked loss)
            # Only calculate loss for non-missing values
            reconstruction_loss = original_dim * binary_crossentropy(y_true * mask, y_pred * mask)
            reconstruction_loss = K.sum(reconstruction_loss, axis=-1) # Sum over features

            # KL divergence loss
            kl_loss = -0.5 * K.sum(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var), axis=-1)

            # VAE loss
            return K.mean(reconstruction_loss + beta * kl_loss) # Mean over batch

        vae = Model(inputs, x_decoded_mean, name='vae')
        vae.add_loss(vae_loss(inputs, x_decoded_mean)) # Add the loss directly to the model
        # Or compile with a dummy loss and rely on model.add_loss:
        # vae.compile(optimizer='adam') # Compile without loss, using model.add_loss

        return vae, decoder # Return both VAE and decoder

    vae, decoder = build_vae(original_dim, latent_dim)

    # Compile with a dummy loss since the actual loss is added via model.add_loss
    vae.compile(optimizer='adam')

    # Train
    # Use X_normalized_original_filled for training input, but the loss uses the mask
    # The input to fit is what y_true in vae_loss will receive
    vae.fit(X, X, shuffle=True, epochs=epochs, batch_size=batch_size, verbose=0)

    encoder_model = Model(vae.input, [vae.get_layer('z_mean').output, vae.get_layer('z_log_var').output])
    z_mean_predicted, z_log_var_predicted = encoder_model.predict(X) # Use the filled data as input

    # Use the mean of z for imputation
    z_predicted = z_mean_predicted # Using mean for reconstruction
    # Alternatively, sample z: z_predicted = sampling([z_mean_predicted, z_log_var_predicted])

    # Decode to get the imputed values
    imputed_normalized = decoder.predict(z_predicted)

    # Only replace the missing values
    imputed_normalized = mask_np * X_normalized_original + (1 - mask_np) * imputed_normalized

    # Denormalize
    imputed_denormalized = imputed_normalized * range_vals + min_val

    # Convert back to original dtype if needed (assuming original was numeric)
    # For categorical data encoded with OrdinalEncoder, we need to inverse transform
    imputed_final = np.round(imputed_denormalized)

    # Inverse transform to get original categorical values
    # Ensure values are within the valid range for each category before inverse transform
    imputed_encoded_clipped = np.copy(imputed_final).astype(int)

    for i in range(imputed_encoded_clipped.shape[1]):
        # Get categories for this column
        categories = encoder.categories_[i]
        # Clip values to be within the valid range of encoded values [0, len(categories)-1]
        imputed_encoded_clipped[:, i] = np.clip(
            imputed_encoded_clipped[:, i],
            0,
            len(categories) - 1
        )


    if np.isnan(imputed_encoded_clipped).any():
        print("Warning: NaN values found after VAE imputation. Falling back to mode imputation for affected columns.")
        temp_df = pd.DataFrame(imputed_encoded_clipped, columns=df_missing.columns)
        for col in temp_df.columns:
            if temp_df[col].isna().any():
                # Get mode from the *original* non-missing data if possible
                original_col_non_missing = df_missing[col].dropna()
                if not original_col_non_missing.empty:
                    mode_val = original_col_non_missing.mode()[0]
                else:
                    # If original column was all NaN, use a placeholder like 0 or a common category
                    # For simplicity, use mode of the (potentially empty) original column,
                    # which might result in NaN, or a default value like 0
                    mode_val = df_missing[col].mode()[0] if not df_missing[col].mode().empty else (categories[0] if len(categories) > 0 else 0)

                temp_df[col] = temp_df[col].fillna(mode_val)
        imputed_final_df = temp_df
    else:
         imputed_final_df = pd.DataFrame(encoder.inverse_transform(imputed_encoded_clipped),
                                        columns=df_missing.columns)


    # Clean up TensorFlow session
    K.clear_session()
    del vae, encoder_model, decoder, X, mask
    gc.collect()
    # Explicitly delete TensorFlow objects
    if 'z_mean' in locals(): del z_mean
    if 'z_log_var' in locals(): del z_log_var
    if 'inputs' in locals(): del inputs
    if 'outputs' in locals(): del outputs
    if 'x_decoded_mean' in locals(): del x_decoded_mean
    if 'decoder_input' in locals(): del decoder_input
    if 'x_decoded' in locals(): del x_decoded
    if 'z' in locals(): del z
    if 'x' in locals(): del x


    return imputed_final_df


def impute_vae_(df_missing, batch_size=32, epochs=100, latent_dim=20, beta=1.0):

    # Encode categorical data
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Normalize data
    min_val = np.nanmin(encoded)
    max_val = np.nanmax(encoded)
    X = (encoded - min_val) / (max_val - min_val)
    X = np.nan_to_num(X, nan=0)

    # Create mask (1: observed, 0: missing)
    mask = (~np.isnan(encoded)).astype(float)

    # Data dimensions
    original_dim = X.shape[1]
    intermediate_dim = 128

    # Encoder network
    inputs = Input(shape=(original_dim,))
    h = Dense(intermediate_dim, activation='relu')(inputs)
    z_mean = Dense(latent_dim)(h)
    z_log_var = Dense(latent_dim)(h)

    # Sampling function
    def sampling(args):
        z_mean, z_log_var = args
        epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent_dim), mean=0., stddev=1.)
        return z_mean + K.exp(0.5 * z_log_var) * epsilon

    # Latent space sampling
    z = Lambda(sampling, output_shape=(latent_dim,))([z_mean, z_log_var])

    # Decoder network
    decoder_h = Dense(intermediate_dim, activation='relu')
    decoder_mean = Dense(original_dim, activation='sigmoid')
    h_decoded = decoder_h(z)
    x_decoded_mean = decoder_mean(h_decoded)

    # VAE model
    vae = Model(inputs, x_decoded_mean)

    # Loss function
    def vae_loss(x, x_decoded_mean):
        reconstruction_loss = original_dim * binary_crossentropy(x, x_decoded_mean)
        kl_loss = -0.5 * K.sum(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var), axis=-1)
        return reconstruction_loss + beta * kl_loss

    # Compile and train
    vae.compile(optimizer='adam', loss=vae_loss)
    vae.fit(X, X, shuffle=True, epochs=epochs, batch_size=batch_size, verbose=0)

    # Impute missing values
    X_imputed = vae.predict(X, verbose=0)

    # Only replace missing values
    X_imputed = mask * X + (1 - mask) * X_imputed

    # Denormalize and convert back to categories
    X_imputed = X_imputed * (max_val - min_val) + min_val
    X_imputed = np.round(X_imputed).astype(int)
    X_imputed = np.clip(X_imputed, 0, len(encoder.categories_[0])-1)

    return pd.DataFrame(encoder.inverse_transform(X_imputed), columns=df_missing.columns)

def impute_bayesian_gmm(df_missing, n_components=5):
    """
    Bayesian Gaussian Mixture Model (GMM) based imputation.

    NOTE: GMM requires complete data for fitting. This implementation
    uses a simple median imputation as an initial step before fitting GMM.
    A more sophisticated approach would integrate GMM into an iterative process.
    """
    if df_missing.empty:
        return df_missing.copy()

    # 1. Ordinal encode categorical data
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # 2. Handle columns that became all NaN after encoding (e.g., columns with only unknown values)
    # Or just handle any remaining NaNs before fitting GMM
    # Use median imputation for simplicity as a first step for the GMM fit
    # Find columns with NaNs
    nan_cols = np.where(np.isnan(encoded).any(axis=0))[0]
    imputed_encoded_initial = encoded.copy()

    for col_idx in nan_cols:
        col_data = encoded[:, col_idx]
        # Calculate median ignoring NaNs
        median_val = np.nanmedian(col_data)
        # If median is NaN (column was all NaN), use 0 or a default value
        if np.isnan(median_val):
             median_val = 0 # Or handle as appropriate, e.g., mode of observed data
        imputed_encoded_initial[:, col_idx] = np.nan_to_num(col_data, nan=median_val)


    # 3. Fit Bayesian GMM on the initially imputed data
    try:
        # Ensure there are enough samples and components are less than samples
        n_samples = imputed_encoded_initial.shape[0]
        n_components = min(n_components, n_samples - 1) # Ensure components < samples

        if n_samples < 2:
             print("Warning: Not enough samples for BayesianGMM. Falling back to median.")
             # Fallback to median imputation if GMM is not feasible
             final_imputed_encoded = np.nan_to_num(encoded, nan=np.nanmedian(encoded, axis=0))
        else:
            gmm = BayesianGaussianMixture(n_components=n_components, random_state=42)
            # Fit on the initially imputed data
            gmm.fit(imputed_encoded_initial)

            # 4. Use the fitted GMM to impute the *original* missing values
            # This is a simplified approach - a full GMM imputation might
            # involve sampling from components based on posterior probabilities.
            # Here, we'll predict the component for each row and use the component mean
            # as the imputation (for missing values only).
            # A more correct approach would be to sample or use conditional expectation.

            # Predict component for each row (using the initially imputed data)
            # Note: predict requires no NaNs
            component_assignments = gmm.predict(imputed_encoded_initial)

            final_imputed_encoded = encoded.copy() # Start with the original encoded data

            for i in range(encoded.shape[0]): # Iterate through rows
                if np.isnan(encoded[i, :]).any(): # Check if the original row had NaNs
                     assigned_component = component_assignments[i]
                     # Get the mean of the assigned component
                     component_mean = gmm.means_[assigned_component]
                     # Impute the missing values in the original row with the component mean
                     missing_mask_row = np.isnan(encoded[i, :])
                     final_imputed_encoded[i, missing_mask_row] = component_mean[missing_mask_row]

            # Ensure no NaNs somehow crept back in (shouldn't happen with this logic)
            final_imputed_encoded = np.nan_to_num(final_imputed_encoded, nan=0) # Final safety net


    except Exception as e:
        print(f"Error during BayesianGMM imputation: {str(e)}. Falling back to median imputation.")
        # Fallback to median imputation if GMM fitting or prediction fails
        final_imputed_encoded = np.nan_to_num(encoded, nan=np.nanmedian(encoded, axis=0))


    # 5. Convert imputed values back to integer indices and clip to valid range
    final_imputed_encoded_int = np.round(final_imputed_encoded).astype(int)

    # Clip to valid category ranges for each column
    for i in range(final_imputed_encoded_int.shape[1]):
        # Ensure the column exists in the encoder categories
        if i < len(encoder.categories_):
            final_imputed_encoded_int[:, i] = np.clip(
                final_imputed_encoded_int[:, i],
                0,
                len(encoder.categories_[i]) - 1
            )
        else:
            # Handle case where encoded had more columns than encoder.categories_
            # This shouldn't happen if encoder was fit on the data used for encoded
            print(f"Warning: Column index {i} out of bounds for encoder categories.")
            final_imputed_encoded_int[:, i] = 0 # Default to 0 or handle as appropriate

    # 6. Inverse transform back to original format
    try:
        imputed_data = encoder.inverse_transform(final_imputed_encoded_int)
        return pd.DataFrame(imputed_data, columns=df_missing.columns)
    except Exception as e:
        print(f"Error during inverse transform: {str(e)}")
        # If inverse transform fails, something is fundamentally wrong with the encoded data/indices
        # Return the raw encoded data or raise an error
        raise RuntimeError(f"Inverse transform failed after GMM imputation: {str(e)}")


def impute_bayesian_gmm_(df_missing, n_components=5):
    """
    Bayesian Gaussian Mixture Model imputation.
    Automatically determines number of clusters.
    """
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Fit model
    gmm = BayesianGaussianMixture(n_components=n_components, random_state=42)
    gmm.fit(encoded[~np.isnan(encoded).all(axis=1)])

    # Predict and impute
    imputed = encoded.copy()
    missing_mask = np.isnan(encoded)

    for i in range(encoded.shape[0]):
        if missing_mask[i].any():
            # Sample from posterior
            sample = gmm.sample(1)[0][0]
            imputed[i, missing_mask[i]] = sample[missing_mask[i]]

    # Convert back
    imputed = np.round(imputed).astype(int)
    imputed = np.clip(imputed, 0, len(encoder.categories_[0])-1)

    return pd.DataFrame(encoder.inverse_transform(imputed), columns=df_missing.columns)

def gru_d_impute(df_missing, epochs=100):
    """GRU-D imputation for missing data"""
    # Encode categorical data
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Normalize data
    min_val = np.nanmin(encoded, axis=0)
    max_val = np.nanmax(encoded, axis=0)
    # Add a small epsilon to avoid division by zero for columns with constant values
    range_vals = max_val - min_val + K.epsilon()

    X_normalized_original = (encoded - min_val) / range_vals

    # Create mask (1: observed, 0: missing)
    mask = (~np.isnan(encoded)).astype(float)

    # Temporarily fill missing with 0 for input to the network
    X_filled = np.nan_to_num(X_normalized_original, nan=0)

    n_samples, n_features = X_filled.shape

    X_input = X_filled.reshape(n_samples, 1, n_features)
    mask_input = mask.reshape(n_samples, 1, n_features)


    # Re-fill NaNs with np.nan for Masking layer
    X_input_nan = X_normalized_original.reshape(n_samples, 1, n_features)
    mask_input_nan = mask.reshape(n_samples, 1, n_features) # Keep mask as is

    # Create input layers
    input_layer = Input(shape=(1, n_features)) # Shape is (timesteps, features)

    # Input layer with Masking (masks values equal to mask_value)
    masked_input = Masking(mask_value=0.)(input_layer)

    # GRU layer
    gru_out = GRU(units=n_features, return_sequences=True)(masked_input) # Output shape (batch_size, timesteps, units)

    outputs = Dense(n_features, activation='sigmoid')(gru_out) # Predict normalized values (0-1 range)

    # Define the model
    model = Model(inputs=input_layer, outputs=outputs)

    # Compile the model
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

    # Prepare training data: input is X_filled, target is X_filled (autoencoder style)
    # X_filled shape is (n_samples, n_features). Reshape for GRU input (n_samples, 1, n_features).
    X_train = X_filled.reshape(n_samples, 1, n_features)
    #Start with: batch_size = 32
    #If training is too slow or memory is underutilized, try: 64 or 128
    #If you're running into OOM (Out of Memory) errors, reduce: 16 or 8

    batch_size = 64

    # Train the model
    model.fit(X_train, X_train, epochs=epochs, batch_size=batch_size, verbose=0)

    # Predict imputed values
    X_imputed_normalized = model.predict(X_train)

    # The prediction has shape (n_samples, 1, n_features). Remove the timestep dimension.
    X_imputed_normalized = X_imputed_normalized.reshape(n_samples, n_features)


    # Only replace missing values using the mask
    imputed_normalized = X_normalized_original.copy() # Start with original normalized data (with NaNs)
    # Fill NaNs with the predicted values
    imputed_normalized[np.isnan(imputed_normalized)] = X_imputed_normalized[np.isnan(imputed_normalized)]

    # Denormalize
    imputed_encoded = imputed_normalized * range_vals + min_val

    # Inverse transform (requires integer categories)
    # Clip values to be within the range of categories for each column
    imputed_encoded = np.round(imputed_encoded).astype(int)

    for i in range(imputed_encoded.shape[1]):
         if i < len(encoder.categories_):
              imputed_encoded[:, i] = np.clip(imputed_encoded[:, i], 0, len(encoder.categories_[i]) - 1)
         else:
             # Handle potential new columns or issues in shape alignment
             print(f"Warning: Column index {i} out of range for encoder categories ({len(encoder.categories_)}). Clipping to [0, 0].")
             imputed_encoded[:, i] = np.clip(imputed_encoded[:, i], 0, 0) # Or handle appropriately


    # Convert back to DataFrame
    imputed_df = pd.DataFrame(encoder.inverse_transform(imputed_encoded), columns=df_missing.columns)

    # Clean up Keras session
    K.clear_session()
    del model, input_layer, masked_input, gru_out, outputs
    gc.collect()

    return imputed_df


def gru_d_impute_(df_missing, epochs=50):
    """GRU-D imputation for time-series missing data."""
    # Preprocess data (fill NaNs with 0 and create a mask)
    data = df_missing.fillna(0).values
    mask = (~df_missing.isna()).astype(int).values

    # Normalize data (critical for GRU-D)
    mean = np.nanmean(data, axis=0)
    std = np.nanstd(data, axis=0)
    data = (data - mean) / std

    # Time steps (assuming each row is a time step)
    n_timesteps, n_features = data.shape

    # GRU-D model
    inputs = Input(shape=(n_timesteps, n_features))
    masked = Masking(mask_value=0)(inputs)  # Mask missing values

    # GRU-D layer (custom decay handling)
    gru_out = GRU(units=16, return_sequences=True)(masked)
    reshaped = Reshape((-1, 64))(gru_out)  # Fixes shape issues
    outputs = Dense(n_features)(reshaped)

    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mse')

    # Train (use data as both input and target)
    model.fit(
        x=data[np.newaxis, ...],  # Add batch dimension
        y=data[np.newaxis, ...],
        epochs=epochs,
        verbose=0
    )

    # Impute missing values
    imputed = model.predict(data[np.newaxis, ...])[0]
    imputed = imputed * std + mean  # Reverse normalization
    return pd.DataFrame(imputed, columns=df_missing.columns)


def impute_grud(df_missing, epochs=50, batch_size=32):
    """
    Gated Recurrent Unit with Decay)
    GRU-D imputation for time-series-like categorical data.
    Implements decay mechanisms for missing values.
    """
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    encoded = encoder.fit_transform(df_missing)

    # Normalize
    min_val = np.nanmin(encoded)
    max_val = np.nanmax(encoded)
    X = (encoded - min_val) / (max_val - min_val)
    X = np.nan_to_num(X, nan=0)

    # Create mask and time delta
    mask = (~np.isnan(encoded)).astype(float)
    time_delta = np.zeros_like(X)  # Simplified - could use actual timestamps

    # Build GRU-D model
    input_dim = X.shape[1]
    inputs = Input(shape=(input_dim,))
    mask_input = Input(shape=(input_dim,))
    delta_input = Input(shape=(input_dim,))

    # Concatenate inputs
    concat = Concatenate()([inputs, mask_input, delta_input])

    gru_out = GRU(64, return_sequences=False)(
    Reshape((1, input_dim * 3))(concat)
)

    # Output layer
    outputs = Dense(input_dim, activation='sigmoid')(gru_out)

    model = Model(inputs=[inputs, mask_input, delta_input], outputs=outputs)
    model.compile(optimizer='adam', loss='mse')

    # Train
    model.fit([X, mask, time_delta], X,
              epochs=epochs,
              batch_size=batch_size,
              verbose=0)

    # Impute
    X_hat = model.predict([X, mask, time_delta], verbose=0)
    X_imputed = mask * X + (1 - mask) * X_hat

    # Denormalize and convert back
    X_imputed = X_imputed * (max_val - min_val) + min_val
    X_imputed = np.round(X_imputed).astype(int)
    X_imputed = np.clip(X_imputed, 0, len(encoder.categories_[0])-1)

    return pd.DataFrame(encoder.inverse_transform(X_imputed), columns=df_missing.columns)


def impute_em(df_missing, max_iter=100, tol=1e-4):
    """
    Robust Expectation-Maximization imputation
    Iteratively estimates parameters (e.g., probabilities) and imputes missing values.
"""
    # Data Encoding
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
    encoded = encoder.fit_transform(df_missing)

    # Initial imputation with mean
    col_means = np.nanmean(encoded, axis=0)
    imputed = np.where(np.isnan(encoded), col_means, encoded)

    n_samples, n_features = encoded.shape
    n_components = min(3, n_features-1)  # Ensure n_components < n_features

    for iteration in range(max_iter):
        # E-step: Dimensionality Reduction
        pca = PCA(n_components=n_components)
        transformed = pca.fit_transform(imputed)

        # M-step: Reconstruction
        reconstructed = pca.inverse_transform(transformed)

        # Update only missing values
        update_mask = np.isnan(encoded)
        imputed[update_mask] = reconstructed[update_mask]

        # Convergence Check
        if iteration > 0 and np.linalg.norm(imputed - prev_imputed) < tol:
            break
        prev_imputed = imputed.copy()

    # Ensure proper categorical values
    imputed = np.round(imputed).astype(int)
    imputed = np.clip(imputed, 0, len(encoder.categories_[0])-1)

    return pd.DataFrame(encoder.inverse_transform(imputed),
                       columns=df_missing.columns)
```

```python
class ImputationRecommender:
    """Complete meta-learning system for imputation method recommendation"""
    def __init__(self):
        self.feature_names_in_training = None
        self.meta_model = None
        self.feature_scaler = StandardScaler()
        self.label_encoder = LabelEncoder()
        self.meta_feature_names = None

    def fit(self, meta_data):

        # Check if target column is present
        if 'best_method' not in meta_data.columns:
            raise ValueError("meta_data must contain a 'best_method' column for training.")

        # Separate features and target
        X = meta_data.drop(columns=['best_method'])
        y = meta_data['best_method']

        # Keep only numeric features
        numeric_X = X.select_dtypes(include=np.number)
        if numeric_X.empty:
            raise ValueError("No numeric features available in meta_data for scaling.")

        self.meta_feature_names = numeric_X.columns.tolist()

        # Scale features
        self.feature_scaler = StandardScaler()
        X_scaled = self.feature_scaler.fit_transform(numeric_X)

        # Encode target labels
        self.label_encoder = LabelEncoder()
        y_encoded = self.label_encoder.fit_transform(y)

        # Count class instances
        class_counts = np.bincount(y_encoded)
        min_class_count = class_counts.min()
        n_classes = len(class_counts)
        n_samples = len(meta_data)

        print(f"Training samples: {n_samples}, Classes: {n_classes}, Min samples/class: {min_class_count}")

        # Choose cross-validation strategy
        if min_class_count >= 2:
            n_splits = min(5, min_class_count)
            print(f"Using StratifiedKFold with {n_splits} splits")
            cv = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
        elif n_samples >= 2:
            print("Not enough samples per class for StratifiedKFold; using KFold with 2 splits")
            cv = KFold(n_splits=2, shuffle=True, random_state=42)
        else:
            raise ValueError("Not enough samples to train the model.")

        # Initialize and train model
        self.meta_model = RandomForestClassifier(n_estimators=200, random_state=42)

        try:
            scores = cross_val_score(self.meta_model, X_scaled, y_encoded, cv=cv)
            print(f"Cross-validation accuracy: {np.mean(scores):.3f} ± {np.std(scores):.3f}")
        except ValueError as e:
            print(f"Cross-validation failed: {e}")

        # Final model training
        self.meta_model.fit(X_scaled, y_encoded)
        print("Model training completed.")

        return self


    def recommend(self,evaluation ,meta_features, new_data, labels, missing_rate, seed, name):
      if self.meta_model is None or self.feature_scaler is None or self.label_encoder is None:
          raise ValueError("Model not trained yet. Call fit() first.")

      # 1. Extract meta-features (includes seed)
     # meta_features = extract_meta_features(new_data, labels, missing_rate,         seed,name)

      # 2. Convert to DataFrame if not already
      if not isinstance(meta_features, pd.DataFrame):
          meta_features = pd.DataFrame([meta_features])

      # 3. Ensure all training features are present
      missing_features = set(self.meta_feature_names) - set(meta_features.columns)
      for feature in missing_features:
          meta_features[feature] = 0  # Fill missing features with default value

      # 4. Remove any extra features not seen during training
      extra_features = set(meta_features.columns) - set(self.meta_feature_names)
      print('extra_features extra_features ',extra_features)
      if extra_features:
          meta_features = meta_features.drop(columns=list(extra_features))

      # 5. Align feature order with training data
      meta_features = meta_features[self.meta_feature_names]

      # 6. Select only numeric features for scaling
      meta_features_numeric = meta_features.select_dtypes(include=np.number)

      # 7. Validate feature count matches training
      if meta_features_numeric.shape[1] != len(self.meta_feature_names):
          raise ValueError(
              f"Feature count mismatch. Expected {len(self.meta_feature_names)} "
              f"features, got {meta_features_numeric.shape[1]}. "
              "Check extract_meta_features output."
          )

      # 8. Scale features
      scaled_features = self.feature_scaler.transform(meta_features_numeric)

      # 9. Make prediction
      predicted_index = self.meta_model.predict(scaled_features)[0]
      recommended_method = self.label_encoder.inverse_transform([predicted_index])[0]


      probas = self.meta_model.predict_proba(scaled_features)

      # 2. Get **all indices** sorted from highest to lowest probability
      top_k_all = np.argsort(probas[0])[::-1]
      print("Indices for top_k (All) from highest to lowest:", top_k_all)
      # 3. Also get the actual method names for all
      methods_all = self.label_encoder.inverse_transform(top_k_all)
      print("Methods ranked from highest to lowest probability:", methods_all)

      return recommended_method, top_k_all, methods_all

    def explain_recommendation(self,evaluation, meta_features_raw, new_data, labels, missing_rate,
                               seed,name, show_plot=True, save_path=None):
      try:
          # Step 1: Extract meta-features
          meta_features_raw = extract_meta_features(new_data, labels,        missing_rate, seed,name)

          # Step 2: Wrap in DataFrame, convert to numeric
          features_df = meta_features_raw.apply(pd.to_numeric, errors='coerce')

          # Step 3: Align features with trained model input
          aligned_features = pd.DataFrame(columns=self.meta_feature_names)
          for col in self.meta_feature_names:
              aligned_features[col] = features_df.get(col, pd.Series([0]))

          # Step 4: Scale the features
          scaled = self.feature_scaler.transform(aligned_features)

          # Step 5: Predict method
          prediction = self.meta_model.predict(scaled)[0]
          predicted_method = self.label_encoder.inverse_transform([prediction])[0]

          # Step 6: SHAP explanation
          explainer = shap.TreeExplainer(self.meta_model)
          shap_values = explainer.shap_values(scaled)

          # Step 7: Handle SHAP output shape
          if isinstance(shap_values, list):
              shap_values = shap_values[prediction]
          elif shap_values.ndim == 3:
              shap_values = shap_values[0, :, prediction]

          # Ensure 2D: (1, n_features)
          shap_values = np.array(shap_values).reshape(1, -1)

          # Step 8: Visualization
          plt.figure(figsize=(10, 5))
          shap.summary_plot(
              shap_values,
              scaled,
              feature_names=self.meta_feature_names,
              plot_type="bar",
              show=False )

          plt.title(f"Recommendation Explanation: {predicted_method}")
          plt.tight_layout()
          if show_plot:
              plt.show()
          if save_path:
              plt.savefig(save_path)

          # Step 9: Return structured explanation
          return {
              'method': predicted_method,
              'features': aligned_features.iloc[0].to_dict(),
              'shap_values': dict(zip(self.meta_feature_names, shap_values.flatten()))
          }

      except Exception as e:
          print(f"Explanation failed: {str(e)}")
          return None
```

```python
def convert_meta_features(meta_features):
    if isinstance(meta_features, pd.DataFrame) and len(meta_features) == 1:
        return meta_features.iloc[0].to_dict()
    elif isinstance(meta_features, np.ndarray) and meta_features.ndim <= 2:
        if meta_features.ndim == 1:
            return {f'feature_{i}': meta_features[i] for i in range(meta_features.shape[0])}
        elif meta_features.ndim == 2 and meta_features.shape[0] == 1:
            return {f'feature_{i}': meta_features[0, i] for i in range(meta_features.shape[1])}
    elif isinstance(meta_features, dict):
        return meta_features
    else:
        try:
            return dict(meta_features)
        except Exception as e:
            print(f"Could not convert meta_features: {e}")
            return {}
```
